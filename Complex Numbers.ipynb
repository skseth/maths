**Theorem: (Mean Value Theorem)** : Given f is continuous on $[a,b]$ and differentiable on (a,b), there is a point c in (a,b) such that :

$f'(c) = \frac{f(b) - f(a)}{b - a}$

In brief, this means there is at least one point c between a and b, where the tangent to f is parallel to the the chord joining (a,f(a)) and (b, f(b)).

**Theorem: If a function f has zero derivative everywhere on an interval, it must be constant on that interval.**  
For any a,b on the interval :  
By mvt $f'(c) = \frac{f(b) - f(a)}{b - a} = 0 => f(b) = f(a)$.

# Inner Product

For $R^n$, we can define an dot product between two vectors. This dot product is used to define a norm (which defined the length of a vector), which can then be used to define a metric, which defined the topology. So this is a key definition.

If $v = (v_1,v_2,...,v_n)$ and $w = (w_1, w_2,...,w_n)$ are vectors, then their dot product is defined as 

$$\langle v, w \rangle = \sum_{i=1}^{n} v_iw_i$$

The definition has the following properties

If 𝑢, 𝑣, 𝑤 are vectors and 𝑐 is a scalar. Then the dot product has the following properties :

$\langle u, v \rangle = \langle v, u \rangle$ - (symmetry)  
$\langle u, av + bw \rangle = a\langle u, v \rangle + b\langle u, w \rangle$ - (linearity - bilinearity with symmetry)   
$\langle v, v \rangle \geq 0$ with 0 only when v = 0. (positive-definiteness)  

However, for complex numbers, the square of a complex number can be negative (e.g. $i^2 = -1). This fact has an impact on the definition of the inner product. e.g. if we used the same axioms as for real numbers, we find :

$\langle iu, iu \rangle = i\langle u, iu \rangle = i^2\langle u, u \rangle = -\langle u, u \rangle$ which implies the norm of one of the vectors is not positive, or both are zero. So we are forced to change our definitions of inner product. Basically, we give up symmetry, and instead opt for conjugate symmetry.

$$\langle v, w \rangle = \sum_{i=1}^{n} v_i\overline{w_i}$$

$\langle u, v \rangle = \overline{\langle u, v \rangle}$ - (conjugate symmetry)  
$\langle u, av + bw \rangle = a\langle u, v \rangle + b\langle u, w \rangle$ - (linearity - but not bilinear)   
$\langle v, v \rangle \geq 0$ with 0 only when v = 0. (positive-definiteness)  

Note that linearity could have been defined for the first or second parameter (because of conjugate symmetry, only one parameter can be linear). 

This definition has several advantages :
- it reduces to the real inner product if the vectors have real coordinates - in this case we get back symmetry and bilinearity.
- it matches the definition of the $|z| = \sqrt{z\bar{z}}$ of a complex number

## Norm, Distance and Angle

Based on this definition, we define the length of a vector :

$$\|v\| = \langle v,v \rangle^{1/2}$$

Distance between two vectors then follows

$$d(v,w) = \|v - w\|$$

Note that d(v,w) = 0 => v - w = 0 => v = w.

The angle between two vectors can be defined as :

$$cos(\theta) = \frac{\langle u,v \rangle}{\|u\|\|v\|}$$

## Pythagorean Theorem

Lemma : $z = a - ib$ and $w = c + id$ are two complex numbers. If $z\bar{w}$ = 0, then $\bar{z}w = 0$  
$z\bar{w} = 0 = (ac + bd) + i(bc - ad)$ => ac = -bd, and bc = ad. Hence  
$\bar{z}{w} = 0 = (ac + bd) + i(-bc + ad) = 0$

Now, we have for any two vectors :
$$\| u + v \|^{2} = \langle u + v, u + v \rangle = \langle u,u \rangle + \langle u,v \rangle + \langle v,u \rangle + \langle v,v \rangle$$

If \langle u,v \rangle = 0 => \langle v,u \rangle = 0, because of the definition of inner product, and the lemma above. 

It follows that when \langle u,v \rangle = 0,

$$\| u + v \|^{2} = \langle u,u \rangle + \langle v,v \rangle = \|u\|^2 + \|v\|^2$$

This is the Pythagorean theorem in n-dimensions, and of course also holds for two dimensions.

## Cauchy Schwarz Inequality

Inner products as defined here satisfy an important inequality :

$$|\langle u,v \rangle| \leq \|u\|\|v\|$$

The equality holds only if u and v are scalar multiples of each other (i.e. are linearly dependent), or one of them is zero. 

The basic statement is that the dot product is smaller than the product of the lengths of the vectors - which in turn implies that a projection of a vector is shorter than the vector itself - a reasonable statement. It also implies that $cos(\theta)$ has a magnitude less than or equal to 1, with equality being when the angle is 0 or $2\pi$

Proof : (i) For v = 0, clearly there is equality since both sides are zero, same is true for u = 0. And a zero vector is always linearly dependent with any other vector. So the theorem is proved for this case.

(ii) Assume v is not 0. Then let 
$$z = u - \frac{\langle u,v \rangle}{\langle v,v \rangle}v$$
z is basically the perpendicular component of u from v. Then

$$\langle z,v \rangle = \langle u - \frac{\langle u,v \rangle}{\langle v,v \rangle}v,v \rangle = \langle u,v \rangle - \frac{\langle u,v \rangle}{\langle v,v \rangle}\langle v,v \rangle = 0$$

Thus z and v are perpendicular. Thus we can apply the Pythagorean theorem to 
$$u = \frac{\langle u,v \rangle}{\langle v,v \rangle}v + z$$

We get, 

$$\|u\|^2 = \frac{\langle u,v \rangle^2}{\|v\|^2} $$, from which the inequality follows.

## Triangle Inequality

$$\|u + v\| \leq \|u\| + \|v\|$$  

Proof: $\|u + v\|^2 = \|u\|^2 + \langle u,v \rangle + \langle v,u \rangle + \|v\|^2$

Now, $\langle u,v \rangle + \langle v,u \rangle \leq 2\|u\|\|v\|$, by the Cauchy-Schwarz inequality.

Substituting, we get the triangle inequality. Now, it follows that :

$\|u + v - v\| \leq \|u + v\| + \|v\| => \|u + v\| \geq \|u\| - \|v\|$. By symmetry, we can see that $\|u + v\| \geq \|v\| - \|u\|$. We conclude that$\|u + v\| \geq |\|v\| - \|u\||$. 

We can write these inequalities more generally as :

$$\|\pm u \pm v\| \leq \|u\| + \|v\| \textbf{  (triangle inequality)}$$

$$\|\pm u \pm v\| \geq |\|u\| - \|v\|| \textbf{  (reverse triangle inequality)}$$

By induction, we get :

$$\sum_{k=1}^{n} v_k \leq \sum_{k=1}^{n} \|v_k\| \textbf{  (triangle inequality for sums)}$$

## Metric Space

A metric space is a set together with metric d defined on it such that :

$d(x,y) = 0$ <=> x = y (distinct points have a non-zero distance)  
$d(x,y) = d(y,x)$ (symmetry)  
$d(x,y) = d(x,z) + d(z,y)$ (triangle inequality)  

The distance defined by the inner product makes $C^n$ and $R^n$ into a metric space.

# Continuity / Differentiability of Paths

Path / Curve - continuous map: $\gamma(t) = [a,b] \to C$ where $[a,b]$ is a closed interval in R.  
smooth path - $\gamma$ is continuous on $[a,b]$, $\gamma'(t)$ differentiable (non-zero) on (a,b)

Continuity : 

$$\lim_{t \to t_0} \gamma(t) = \gamma(t_0)$$

Derivative of the curve: 
$$\gamma'(t_0) = \lim_{t \to t_0} \frac{\gamma(t_0) - \gamma(t)}{t - t_0}$$  

Every path has an orientation 

A path is simple (does not cross itself) if $\gamma(t)$ is one-one, except possibly $\gamma(a) = \gamma(b)$. A path $\gamma$ is closed if $\gamma(a) = \gamma(b)$. 

E.g. Unit Circle $e^{it}$ is a simple parameterized path.

Any path is connected (see topology)  

Path-connectedness => connectedness

If G is connected, then between any two points in G there exists a path entirely in G. Also if G is open and connected, we can form a path for any two points compaosed only of horizontal and vertical components lying entirely in G (a **chain of segments**)

Proof: 
Lemma : Given an open disk D centered on (x0,y0) with radius r , given any two points (x1,y1) and (x2,y2) we can find a "horizontal" and "vertical" path entirely within D. Assume (x1,y2) is in D. Then the path (x1,y1) -> (x1, y2) -> (x2,y2) is entirely in D. Assume (x1, y2) is not in D => (x1 - x0)^2 + (y2 - y0)^2 >= r. Since (x2 - x0)^2 + (y2-y0)^2 < r => (x2 - x0)^2 < (x1-x0)^2 => (x2 - x0)^2 + (y1 - y0)^2 < r. Thus the point (x2, y1) is in D, and the proof follows.

Now a path is a continuous image of a segment $[a,b]$, which is compact - continuous image of a compact set is compact. Given a path between any two points in G, we can cover this path with open disks entirely in G, and such a path has a finite subcover since the path is compact. Now, for any two intersecting disks, choose a point in their intersection, plus choose the the end points f(a) and f(b) (f being the mapping for the path). Then, we have a finite number of points in the path order them by increasing value of the parameter. Each successive pair of points must be in a common open disk from the finite cover. e.g. given f(a) either there is a point (at least f(b)) - other wise there is an open set in the open cover disjoint from the rest, and then we can see that the path is covered by disjoint open sets i.e. is disconnected, which is not possible. So the open set containing f(a) must contain another point in another intersecting disk. We can then break this into a horizontal and vertical section, and repeat this by induction for all points selected from the intersections.

# Differentiation

**Definition (Differentiable)** Suppose f: G -> C is a complex function, and $z_0$ is an interior point of G. The **derivative** of f at $z_0$ is defined as :

$$f'(z_0) = \lim_{z \to z_0} \frac{f(z) - f(z_0)}{z - z_0}$$

OR  

$$f'(z_0) = \lim_{h \to 0} \frac{f(z_0 + h) - f(z_0)}{h}$$


provided the limit exists. In this case, f is said to be **differentiable** at $z_0$. 

**Definition (Holomorphic)** A function f is said to be holomorphic at $z_0$, if there is an open disk centered at $z_0$ where f is differentiable at each point. If a function f is differentiable at every point in an open set E (and hence holomorphic at every point in E), then f is holomorphic in E. A function holomorphic on C is called **entire**.


**Thoerem: If f is differentiable at $z_0$, f is continuous.**  

**Theorem: Derivatives - common properties: Complex functions derivatives under addition, product, division, and composition are the same as for real functions**  


**Thoerem: Derivative of $z^n = nz^{n-1}$**  

**Theorem: A function which is holomorphic at a point a with a non-zero derivative, it preserves angles between smooth paths intersecting at a. A function that preserves angles of intersecting paths is called conformal.**   

Let $\gamma_1(t)$ and $\gamma_2(t)$ be a paremeterization of the paths such that both are at a, for t = 0. 

Since f is holomorphic at a, consider a restriction of f in an open disk E around a, where f is differentiable at every point.

The derivative $g'(\gamma_1(0)) = g'(a)\gamma_1'(0)$ and $g'(\gamma_2(0)) = g'(a)\gamma_2'(0)$. Both the path derivatives exist as the paths are smooth and g'(a) = f'(a) exists and is non-zero by hypothesis. Since multiplying by a non-zero complex number rotates and scales straight lines but does not change the angle between them, the two tangent lines formed by $\gamma_1'$ and $\gamma_2'$ have the same angle between them as $f'(a)\gamma_1'(0)$ and $f'(a)\gamma_2'(0)$.

**Theorem: Derivative of inverse** f:H->G is a function, with an inverse g (thus f is bijective i.e. one-one and onto). G,H are open sets in C. If f is differentiable (non-zero) at g(z_0), and g is continuous at $z_0$, then g is differentiable at $z_0$ and 

$$g'(z_0) = \frac{1}{f'(g(z_0))}$$

Proof: Set let $w_0 = g(z_0)$. Also $\phi(w) = \frac{f(w) - f(w_0)}{w - w_0}$ when $w \neq w_0$, and $\phi(w_0) = f'(w_0)$. $\phi$ is continuous at $w_0$ because f is differentiable at $w_0$. We can show $g'(z_0) = \lim_{z \to z_0} \frac{1}{\phi(g(z))}$. Then, because $\phi$ is continuous we can take the limit inside. So :

$$g'(z_0) = \frac{1}{\phi(\lim_{z \to z_0}g(z))} = \frac{1}{\phi(g(z_0))} = \frac{1}{f'(g(z_0))}$$

Note: We can write a function f(z) = f(x,y) = u(x,y) + iv(x,y), where z = x + iy. 

**Theorem (Cauchy-Riemann Equation): Suppose f is differentiable at $z_0 = x_0 + iy_0$. Then the partial derivatives of f exist and satisfy 
$$f'(z_0) = \frac{\partial f}{\partial x} = -i\frac{\partial f}{\partial y}$$**. 


Proof: Result follows almost immediately by considering $\Delta z = \Delta x$, and $\Delta z = i\Delta y$ in the derivative definition. Since f is differentiable at $z_0$, both limits of the difference quotient must exist and be equal to the derivative.

**Corollary: If f is differentiable at $z_0$, and the second partial derivatives exist, f is harmonic.** 
Proof: 
By the previous rule :  
$$f'(z_0) = u_x + iv_x = v_y - iu_y$$**  
Then, $u_x = v_y$ and $v_x = -u_y$. Thus, taking second derivatives, $u_xx = v_yx = v_xy = -u_yy$ => $u_xx + u_yy = 0$. Thus f is harmonic.


**Theorem: Suppose a function f has partial derivatives $f_x$ and $f_y$ satisfying the Cauchy-Riemann equations. And these partial derivatives are continuous in an open disk centered on $z_0$. Then f is differentiable at $z_0$, and the derivative is equal to $f_x(z_0)$**  

Proof: We have to prove $f'(z_0) = f_x(z_0)$. Using $\Delta z = \Delta x + i\Delta y$ and the Cauchy-riemann condition, we get :

$$f_x(z_0) = \frac{\Delta x}{\Delta z}(f_x(z_0)) + \frac{\Delta y}{\Delta z}(f_y(z_0))$$

Also, using the definition of derivative, the previous equations and some jugglery, we get :

$$f'(z_0) - f_x(z_0) = \lim_{z \to z_0} \frac{\Delta y}{\Delta z}\big (\frac{f(z_0 + \Delta x + i\Delta y) - f(z_0 + \Delta x)}{\Delta y} - f_y(z_0) \big ) + \lim_{z \to z_0} \frac{\Delta x}{\Delta z}\big (\frac{f(z_0 + \Delta x) - f(z_0)}{\Delta x} - f_x(z_0) \big )$$ 

Note that $\Delta x/\Delta z$, and $\Delta y/\Delta z$ are obviously bounded (in absolute value) between 0 and 1.

The right hand part of the sum can be easily evaluated to zero, since $z \to z_0$ => $\Delta x \to 0$
$$\lim_{z \to z_0} \frac{f(z_0 + \Delta x) - f(z_0)}{\Delta x} = \lim_{\Delta x \to 0} \frac{f(z_0 + \Delta x) - f(z_0)}{\Delta x} = f_x(z_0)$$

For the left hand side, we use the mean-value-theorem for real functions : Assume f(z) = f(x,y) = u(x,y) + iv(x,y). Then, by mvt there are real numbers 0 < a,b < 1, such that 

$$\frac{u(x_0 + \Delta x, y_0 + \Delta y) - u(x_0 + \Delta x, y_0)))}{\Delta y} = u_y(x_0 + \Delta x, y_0 + a \Delta y)$$

$$\frac{v(x_0 + \Delta x, y_0 + \Delta y) - v(x_0 + \Delta x, y_0)))}{\Delta y} = v_y(x_0 + \Delta x, y_0 + b \Delta y)$$

$\lim_{\Delta z \to 0} u_y(x_0 + \Delta x, y_0 + a \Delta y) = u_y(x_0, y_0)$ and similarly for v. 

Using this and substituting in the left side, we get the left hand side as zero as well.

**Theorem: If $G \subset C$ is a region ( an open connected subset of C), and f:G -> C is a complex-valued function with f'(z) = 0 for all values in G, then f(z) is constant.**  

Proof: Let $f(z) = f(x,y) = u(x,y) + iv(x,y)$. Between any two points in G we can form a path composed entirely of a finite number of horizontal and vertical segments. Along a horizontal segment H, the y coordinate is constant, say $y_0$. At any point $z_0$ in H, where $z_0 = x_0 + iy_0$,

$f'(z_0) = f_x(z_0) = u_x(x_0,y_0) + iv_x(x_0,y_0) = 0$ => $u_x(x_0,y_0) = 0$ and $v_x(x_0,y_0) = 0$

Since y coordinate is constant, $u'(x_0, y_0) = u_x(x_0,y_0) = 0$ and similarly for v. Thus, by m.v.t. for real functions, both u and v are constant on H. Similarly we can prove this for vertical segments. 

This means given (a horizontal or vertical segment) $[z_0,z_1]$, $f(z_0) = f(z_1)$. Since any two points can be connected by a finite no of horizontal and vertical segments, all points along the whole path have the same value i.e. f is constant.

Note: The need to use horizontal and vertical path arises because we are leaning on using the real m.v.t.

Consider f = u + iv, where at z = x + iy, f(z) = 0 if xy = 0, and 1 otherwise. The function satisfies Cauchy Riemann but fail to be continuous. This is tricky because the partial derivatives exist at 0.

But along the real axis, as we approach point c, the values are 1 on either side of (0,c), but 0 at c. Thus u is not continuous along the x direction (or actually any direction but the y or imaginary line), and $u_x$ cannot exist.

## Mobius Transformations

A **linear fractional transformation** is a function of the form 

$$f(z) = \frac{az + b}{cz + d}$$

Where a,b,c,d are in C.

When ad - bc != 0, the transformation is called a mobius transformation.
