# Arguments

Logic is the study of arguments.

An argument is a set of statements or **propositions** - a series of **premises**, followed by a **conclusion**. A statement (or a declarative statement) is a sentence which can have only one of two **truth values** : true and false.



Sentences that are not propositions :

- questions : where is mumbai?
- proposals/suggestions : let's make a sandwich
- commands : close the door
- exclamations : fantastic!


Passages that are not arguments - Non-inferential arguments
- warnings
- a piece of advise
- opinion / statement of belief
- loosely associated statements
- report
- expository passages
- illustrations
- explanations
- conditional statements

Explanations in particular often look like arguments. "The pipe burst because of the increase in pressure." sounds like an argument. But in fact explanations explain why something is the case, not to prove that it is in fact the case. The statement above is not trying to demonstrate the pipe has burst. It is trying to explain, given the pipe has already burst, why it happened.

An illustration may also look like an argument, because of words like "thus". For example : "Aerobic exercises provide cardiovascular conditioning. Thus, cycling, walking, running all provide this benefit." 

## Conditional Statements and Arguments

A statement like : "if wishes were horses, beggars would ride", is a conditional. They are not arguments, They can be converted to arguments e.g.

Wishes are horses.  
Hence, beggars ride.  

While a conditional statement is not an argument, it can be used in an argument :

if a country has nuclear weapons, it is secure.  
North Korea is a country with nuclear weapons.  
Hence, North Korea is secure.  

A conditional statement can also be a conclusion. 

if the fed reduces rates, funds will move to debt from equity  
if funds move to debt from equity, the stock market will fall  
Hence, if the fed reduces rates, the stock market will fall



Given a conditional : if A then B.

A is a **sufficient condition** for B to be true.  
B is a **necessary condition** for A to be true, i.e. if B is false, so is A. if not B, then not A.  

## Deductive vs Inductive Arguments

Deductive Arguments following "necessary" reasoning, whereas inductive arguments follow "probabilistic" reasoning. Another way to state this is that inductive arguments are *risky*. The vast majority of arguments we work with every day are inductive.

An example of a deductive argument :

All scientists are forgetful.  
Einstein is a scientist.  
Therefore, (necessarily) Einstein is forgetful.  

An example of an inductive argument :

It rains almost every day during monsoon season.  
We are in the monsoon season.    
It will probably raine today.

Historically, deductive arguments were considered to flow from general to the particular, and inductive arguments flowing from the particular to the general. However, this suffers from defects, as many deductive arguments can flow from the particular to the general, and similarly we can find exceptions in the inductive case.

Scientific arguments can be deductive or inductive. When discovering new laws, the reasoning applied is almost always inductive - we look for patterns in nature and come up with a theory which explains this patterns the best. But we cannot rule out the possibility that new information may overturn our theory. However, while applying an agreed upon theory to a new situation, the reasoning is usually deductive.



## Deductive Argument Forms

**argument for mathematics**  

Arguments in mathematics are usually deductive. A notable exception is statistics.

**argument from definition**  

> Mammals are animals with mammary glands  
> Rabbits are mammals  
> Hence rabbits have mammary glands  

A *syllogism* is an argument consisting of exactly two premises and one conclusion.  

**categorical syllogism**  

A categorical syllogism is a syllogism where each statement begins with all, some or no. It is used to make an argument based on properties of a category of objects.

> All trees are plants  
> Some trees are tall  
> Therefore, some plants are tall  

**hypothetical syllogism**  

A hypothetical syllogism is in the form of if ... then ... statements for one or both premises. It is used to make an argument under a given condition or "hypothesis".

> if x then y  
> if y then z  
> hence, if x then z

(a.k.a modus ponens)  
> if x then y  
> x is true  
> hence, y


**disjunctive syllogism**  

> Either x or y  
> x is false  
> y is true

An deductive argument is **valid** if the premises logically imply the conclusion i.e. the conclusion is the logical consequence of the premises. An argument is **sound** if it is valid, and the *premises are true*. The **conclusion** of a sound argument is always **true**. An argument that is not sound, is **unsound**, and an argument that is not valid is **invalid**.

As a general matter soundness of an argument may not always be easy to determine - it may be hard to prove the truth / falsity of the premises.

*Validity of an argument vs the truth of the conclusion*

An argument being valid does not imply the conclusion is true, because the premises could be false. Thus validity does not imply the truth of the conclusion. And invalidity does not imply the conclusion is false.

It is possible for the premises to be true, the conclusion to be true but the argument is still invalid because the premises do not imply the conclusion.

The one impossibility is to have a valid argument with true premises and a false conclusion. This is a key point in deductive logic. When an argument is valid, true premises always lead to true conclusions and the argument is sound.

*superfluous premises*

Addition of a superfluous premise to an otherwise sound argument does not make it unsound - the superfluous premise should not be considered part of the argument.

## Inductive Argument Forms

The forms listed below can overlap.

**prediction**

Proceeds from knowledge of past to make a claim about the future.

**argument from analogy** 

Argument based on analogy between two things or events. We should however be careful that all comparisons are not analogies - e.g. in geometry we may compare two triangles for similarity to conclude the value of an angle. This is a deductive argument as the conclusion is necessary.

**generalization**

Proceed from knowledge of selected sample to some claim about the whole group.

**argument from authority**

An argument that an expert has claimed something, so it must be true.

**argument based on signs**

An argument based on a sign, e.g. a road sign. Here a sign means a message from an intelligent being.

**causal inference**

An argument that proceeds from knowledge of a cause to a claim about an effect.

A **strong inductive argument** is an inductive argument in which it is improbable that the conclusion is false given that the premises are true. A **weak inductive argument** is one where the conclusion does not follow probably from the premises. A **cogent** argument is one which is strong and has all true premises. An **uncogent** argument is one which is not cogent.

All inductive arguments rely on an assumption about uniformity of nature across time / space. Thus past can be predictive of the future, or if something happens in one place it would be happen similarly in a different place.

The strength or weakness of an inductive argument depends on the probabilistic support the premises give to the conclusion. 

One key difference between deductive and inductive arguments is this : If the conclusion of a deductive argument is true independent of the premises, we consider the argument valid (the premises are superfluous). But if the conclusion of a deductive argument is probably true independent of the premises, we consider the argument weak.

A second key difference is that validity in deductive arguments is binary. Strength of deductive arguments is a range. 

The third difference is that addition of more premises to an argument may either strengthen or weaken it, while we have already seen that in a sound deductive argument, adding additional premises has no effect.

On the otheer hand, like deductive arguments, for inductive arguments also determining the truth of premises is not always possible.

An example of an inductive argument :

> Almost all tests conducted so far show the drug result in remission of cancer.  
> Hence, with a high degree of probability, the drug is effective.  

Here if the premise is true, the conclusion is probably true and the argument is strong. But in evaluating the premise it is important to consider the complete evidence. If an unethical scientist has been removing invalid failure results spuriously then we will reach the wrong conclusion. Hence we have the **total evidence requirement** - the expectation is that all relevant evidence is considered to arrive at the conclusion the premise is true.



### Samples and Populations, Proportions and Probability

**Oranges (Generalization)**

> Buying oranges at a grocer, you want to buy a box but check that they are good.  
> The grocer takes an orange from the box, cuts it and shows you it is good.  
> You conclude all the oranges in the box are probably good

This is a risky argument.

You can change the second step to choose the orange yourself. Now, the risk may go down further from earlier.

You may pick six oranges and one is bad. You conclude the box is still worth buying at half-price.

All the arguments above are risky to varying degrees.

Many forms of risky argument involves picking a **sample** from a **population**.

> sample -> population  
> population -> sample  
> sample -> new sample  

By making it more exact, we can use **proportions** to quantify the risk better

> e.g. defining "almost all" by 90%  
> I pick 4 oranges from a box of 60, and see they are good (sample)  
> Hence, almost all oranges in the box (90% = 54) are good (population)  

This leads us to **probability**

> I pick 4 oranges from a box of 60, and see they are good (sample)  
> Hence, probably, almost all oranges in the box (90% = 54) are good (population)  


### Probabilities and risky arguments

Inductive arguments use probabilities. But not all use of probabilities is "risky" - and hence inductive.

Here is an example of a deductive argument using probability :

> A die has 6 faces numbered 1 to 6  
> Each face is equally probable (to turn up when rolled)  
> The probability of rolling a 4 is 1/6  

This is a valid, deductive argument. 

The following is a risky, inductive one :

> A die (not necessarily fair ) has 6 faces numbered 1 to 6  
> In 227 rolls, 4 came up 38 times  
> The probability of rolling a 4 is about 1/6

The conclusion may be false, even with true premises.


## Other types of arguments
### Inference to the best explanation

So far we have seen two types of arguments. In the deductive argument, we argue via necessary reasoning to a firm conclusion. In an inductive argument, we argue via probabalistic reasoning to a probable conclusion. Consider the following inductive argument :

> We picked 100 sample murders from over 1000 murders in City X  
> In 70 of these cases the murderer left the city  
> Hence, it is probable that a murderer leaves the city after committing the crime  

On the other hand, consider a specific murder happening in City X. We have suspects A,B and C. But C has left town soon after the murder. A detective may reason as follows :

> C left town immediately after the murder, while A and B did not  
> Murderers often leave town after a murder  
> Hence, the best explanation is that C committed the crime  

The reasoning in this case is called **abduction**. In induction we are often generalizing from a sample, in abduction we are trying to find the best explanation for a particular case (in this case, a specific murder). We may often apply a result previously obtained via induction to arrive at this best explanation.

### Decision Theory

Arguments can be about changing beliefs. But we also want to take decisions, leading perhaps to actions. 

Decisions are based on 

* what we believe (which we are convinced of by arguments based on reason and evidence, ideally) and
* what we want (values)

We measure values by what are called **utilities**. In economics for example, the value of a product may depend on its utility to us, and may help us in making the decision to purchase something.
