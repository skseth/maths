# Properties of R


We construct R from Q - each member of R is a "cut". A cut is a non-empty proper subset C of Q, such that 
a) if p is in C, and q < p, then q is in C. b) C has no largest member. Using this construction, we identify each point in R with a cut. We can then show that R is an ordered field, and has the least upper bound property. By association each member of Q with a cut, we can show that R contains Q as a sub-field.

R has the order and metric topology and they are equivalent. 
R is a Hausdorff space - any two points can be separated by disjoint open sets.

**The Dedekind Completeness Property or the Least Upper Bound Property**: Suppose S is a nonempty subset of R that is bounded above. Then S has a least upper bound.

By the Heine-Borel Theorem, any closed (bounded) interval \[a,b\] in R is compact. This result depends on the fact that R is an ordered set with the least upper bound property (but not on dense order).

**The Archimedean Property**: For every x,y in R, with x > 0, there is an integer n such that nx > y.

Proof: We can show that if there was no such integer n, then we could create a sequence of numbers in R which are bounded above by y, and hence have a least upper bound, say a. Then, (a - x) < mx for some m => a < (m+1)x, a contradiction, since a is an upper bound of all nx.

**Q is dense in R. Hence R has dense ordering.** : If x < y in R, then there is a $p \in Q$, such that x < p < y. Thus given any two real numbers x and y, there is always a real number between them.
Proof:
y - x > 0.  
n(y - x) > 1, for some n (using Archimedean Property)
-m1 < nx < m2, for some m1, m2, (using Archimedean Property)
m - 1 < nx < m, for some m, such that -m1 <= m <= m2 (only finite no of integers bet -m2 and m1)
Hence, nx < m < 1 + nx < ny => x < m/n < y

Thus Q is dense in R. It follows that given any two distinct real numbers x and y, there is always a real number between them. 

Note that the other definition of dense (topology) is that a set is dense in a space if its closure is the entire space. For Q, every point in R has a neighborhood containing a point of Q (by the above proof). Hence Q is dense in R in that sense as well.

**For every real x > 0, and integer n > 0, there is one and only one real y such that $y^n = x$. If a and b are positive real numbers and n is a positive integer, then $(ab)^{1/n} = a^{1/n}b^{1/n}$**

**The Cut Property** Suppose A and B are nonempty disjoint subsets of R whose union is all of R, such that every element of A is less than every element of B. Then there exists a "cutpoint" c of R, such that every x <= c for every x in A, and y >= c for every y in B.

Proof: x <= sup A <= y for every x in A and every y in B. sup A exists by the least upper bound property.
    
    
**Topological Connectedness**: There is no separation of R i.e. no two non-empty, disjoint open sets whose union is R.  

In fact every interval, and ray in R is connected. This follows immediately from the fact that R is a linear continuum, and hence is connected, as proved in the section on connectedness.

Proof (Alternative): If there was such a partition into two open sets A and B - these are also closed. Choose a from A and b from B, such that a < b (wlog). Then, Let S = {x : x in A and \[a,x\] is in A}. S is bounded above by b, so it has a supremum s (**least-upper-bound property**), which must be in A because A is closed, and the supremum of a set belongs to its closure (because of order topology).

In fact $s \in S$, because otherwise \[a,s\] has some intersection with B containing a point t. (We have to show there are no points of S between t-e and s, leading to a contradiction that s is the supremum.) Since B is open, the interval (t - e, t + e) of this intersection is entirely in B. But this means a < t-e (since a is not in B), and t - e < t < s. So if s the supremum, there must exist an x in S, such that t - e < t+e <= x < s. This means that x is in a, and \[a,x\] is entirely in A, but this is not possible given t is not in A, and t lies between a and x. Thus s must be in S.

But then, since A is open, (s - d, s + d) must be in A. Since R has dense order, there must be a point z of A between s and s + d. But then, since \[a,z\] = \[a,s\] U (s,z] is entirely in A, z belongs to S and s cannot be the supremum.

**The Intermediate Value Property** If f is a continuous function from \[a, b\] to R, with f(a) > 0 and f(b) < 0, then there exists c in (a,b) with f(c) = 0. 

Proof: A = \[a,b\] is connected, as R is a linear continuum. And R is ordered. So ,since f(a) > 0 > f(b), by the Intermediate Value Theorem there must be a point c in \[a,b\], such that f(c) = 0. If it isn't then we would have a separation of f(A) and the image of a connected space is connected under a continuous function.

**The Bounded Value Property**: If f is a continuous function from \[a, b\] to R, there exists B in R with f(x) ≤ B for all x in \[a,b\].

Proof: A = \[a,b\] is compact. Since f is continuous, f(A) is compact in R. Thus f(A) is also closed and bounded.

**The Extreme Value Property**: If f is a continuous function from \[a, b\] to R, there exists c in \[a,b\] with f(x) ≤ f(c) for all x in \[a,b\] (similar is true for minimum). 

Proof: A = \[a,b\] is compact, and f is continuous, R is ordered, so by the the extreme value theorem, f(A) achieves its maximum and minimum. Simply, image f(A) is compact and any compact set in an ordered set has a maximum and minimum element.

**The Shrinking Interval Property**: Suppose I1 ⊇ I2 ⊇ . . . are bounded closed intervals in R with lengths decreasing to 0. Then the intersection of the In’s is nonempty.

**The Nested Interval Property**: Suppose I1 ⊇ I2 ⊇ ... are bounded closed intervals in R. Then the intersection of the In’s is nonempty.

Proof: Both the properties above follow from the theorem that a collection of compact sets having the finite intersection property must have a non-empty intersection (an empty intersection would imply an open cover with no finite subcover). 

**The Convergence of Bounded Monotone Sequences**: Every monotone increasing (or decreasing) sequence in R that is bounded converges to some limit.

**The Convergence of Cauchy Sequences**: Every Cauchy sequence in R is convergent.

**The Fixed Point Property for Closed Bounded Intervals**: Suppose f is a continuous map from \[a, b\] ⊂ R to itself. Then there exists x in \[a, b\] such that f(x) = x. 

**The Contraction Map Property**: Suppose f is a map from R to itself such that for some constant c < 1, |f(x) − f(y)| ≤ c|x − y| for all x,y. Then there exists x in R such that f(x) = x.

**De Moivre's Theorem**  

<div class="alert alert-info">
$$(cos(x) + isin(x))^n = cos(nx) + isin(nx)$$
    
$$(e^{i\theta})^n = e^{in\theta}$$
</div>

Proof: Obviously this is true for n = 1. Assume this is true for n-1. Now, for n, 

$(cos(x) + isin(x))^n = (cos(x) + isin(x))^{n-1}(cos(x) + isin(x)) = (cos((n-1)x) + isin((n-1)x))(cos(x) + isin(x)) $

Let A = (n-1)x and B = x

$cos(A)cos(B) - sin(A)sin(B) = cos(A + B) = cos(nx)$  
$cos(A)sin(B) + sin(A)cos(B) = sin(A + B) = sin(nx)$  
Thus the identity follows.

**Roots of Unity**

# Inner Product

For $R^n$, we can define an dot product between two vectors. This dot product is used to define a norm (which defined the length of a vector), which can then be used to define a metric, which defined the topology. So this is a key definition.

If $v = (v_1,v_2,...,v_n)$ and $w = (w_1, w_2,...,w_n)$ are vectors, then their dot product is defined as 

$$\langle v, w \rangle = \sum_{i=1}^{n} v_iw_i$$

The definition has the following properties

If 𝑢, 𝑣, 𝑤 are vectors and 𝑐 is a scalar. Then the dot product has the following properties :

$\langle u, v \rangle = \langle v, u \rangle$ - (symmetry)  
$\langle u, av + bw \rangle = a\langle u, v \rangle + b\langle u, w \rangle$ - (linearity - bilinearity with symmetry)   
$\langle v, v \rangle \geq 0$ with 0 only when v = 0. (positive-definiteness)  

However, for complex numbers, the square of a complex number can be negative (e.g. $i^2 = -1). This fact has an impact on the definition of the inner product. e.g. if we used the same axioms as for real numbers, we find :

$\langle iu, iu \rangle = i\langle u, iu \rangle = i^2\langle u, u \rangle = -\langle u, u \rangle$ which implies the norm of one of the vectors is not positive, or both are zero. So we are forced to change our definitions of inner product. Basically, we give up symmetry, and instead opt for conjugate symmetry.

$$\langle v, w \rangle = \sum_{i=1}^{n} v_i\overline{w_i}$$

$\langle u, v \rangle = \overline{\langle u, v \rangle}$ - (conjugate symmetry)  
$\langle u, av + bw \rangle = a\langle u, v \rangle + b\langle u, w \rangle$ - (linearity - but not bilinear)   
$\langle v, v \rangle \geq 0$ with 0 only when v = 0. (positive-definiteness)  

Note that linearity could have been defined for the first or second parameter (because of conjugate symmetry, only one parameter can be linear). 

This definition has several advantages :
- it reduces to the real inner product if the vectors have real coordinates - in this case we get back symmetry and bilinearity.
- it matches the definition of the $|z| = \sqrt{z\bar{z}}$ of a complex number

## Norm, Distance and Angle

Based on this definition, we define the norm of a vector :

$$\|v\| = \langle v,v \rangle^{1/2}$$

Distance between two vectors then follows

$$d(v,w) = \|v - w\|$$

Note that d(v,w) = 0 => v - w = 0 => v = w.

The angle between two vectors can be defined as :

$$cos(\theta) = \frac{\langle u,v \rangle}{\|u\|\|v\|}$$

## Pythagorean Theorem

Lemma : $z = a - ib$ and $w = c + id$ are two complex numbers. If $z\bar{w}$ = 0, then $\bar{z}w = 0$  
$z\bar{w} = 0 = (ac + bd) + i(bc - ad)$ => ac = -bd, and bc = ad. Hence  
$\bar{z}{w} = 0 = (ac + bd) + i(-bc + ad) = 0$

Now, we have for any two vectors :
$$\| u + v \|^{2} = \langle u + v, u + v \rangle = \langle u,u \rangle + \langle u,v \rangle + \langle v,u \rangle + \langle v,v \rangle$$

If \langle u,v \rangle = 0 => \langle v,u \rangle = 0, because of the definition of inner product, and the lemma above. 

It follows that when \langle u,v \rangle = 0,

$$\| u + v \|^{2} = \langle u,u \rangle + \langle v,v \rangle = \|u\|^2 + \|v\|^2$$

This is the Pythagorean theorem in n-dimensions, and of course also holds for two dimensions.

## Cauchy Schwarz Inequality

Inner products as defined here satisfy an important inequality :

$$|\langle u,v \rangle| \leq \|u\|\|v\|$$

The equality holds only if u and v are scalar multiples of each other (i.e. are linearly dependent), or one of them is zero. 

The basic statement is that the dot product is smaller than the product of the lengths of the vectors - which in turn implies that a projection of a vector is shorter than the vector itself - a reasonable statement. It also implies that $cos(\theta)$ has a magnitude less than or equal to 1, with equality being when the angle is 0 or $2\pi$

Proof : (i) For v = 0, clearly there is equality since both sides are zero, same is true for u = 0. And a zero vector is always linearly dependent with any other vector. So the theorem is proved for this case.

(ii) Assume v is not 0. Then let 
$$z = u - \frac{\langle u,v \rangle}{\langle v,v \rangle}v$$
z is basically the perpendicular component of u from v. Then

$$\langle z,v \rangle = \langle u - \frac{\langle u,v \rangle}{\langle v,v \rangle}v,v \rangle = \langle u,v \rangle - \frac{\langle u,v \rangle}{\langle v,v \rangle}\langle v,v \rangle = 0$$

Thus z and v are perpendicular. Thus we can apply the Pythagorean theorem to 
$$u = \frac{\langle u,v \rangle}{\langle v,v \rangle}v + z$$

We get, 

$$\|u\|^2 = \frac{\langle u,v \rangle^2}{\|v\|^2} $$, from which the inequality follows.

## Triangle Inequality

$$\|u + v\| \leq \|u\| + \|v\|$$  

Proof: $\|u + v\|^2 = \|u\|^2 + \langle u,v \rangle + \langle v,u \rangle + \|v\|^2$

Now, $\langle u,v \rangle + \langle v,u \rangle \leq 2\|u\|\|v\|$, by the Cauchy-Schwarz inequality.

Substituting, we get the triangle inequality. Now, it follows that :

$\|u + v - v\| \leq \|u + v\| + \|v\| => \|u + v\| \geq \|u\| - \|v\|$. By symmetry, we can see that $\|u + v\| \geq \|v\| - \|u\|$. We conclude that$\|u + v\| \geq |\|v\| - \|u\||$. 

We can write these inequalities more generally as :

$$\|\pm u \pm v\| \leq \|u\| + \|v\| \textbf{  (triangle inequality)}$$

$$\|\pm u \pm v\| \geq |\|u\| - \|v\|| \textbf{  (reverse triangle inequality)}$$

By induction, we get :

$$\sum_{k=1}^{n} v_k \leq \sum_{k=1}^{n} \|v_k\| \textbf{  (triangle inequality for sums)}$$

## Metric Space

A metric space is a set together with metric d defined on it such that :

$d(x,y) = 0$ <=> x = y (distinct points have a non-zero distance)  
$d(x,y) = d(y,x)$ (symmetry)  
$d(x,y) = d(x,z) + d(z,y)$ (triangle inequality)  

The distance defined by the inner product makes $C^n$ and $R^n$ into a metric space.
