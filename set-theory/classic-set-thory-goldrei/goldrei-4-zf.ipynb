$\forall v \phi$, for any variable v

## Axioms 1-3: Extensionality, Empty Set, Pairs

### ZF1: Axiom of Extensionality

$\forall x \forall y(x = y \iff \forall z(z \in x \iff z \in y))$

Two sets are equal if and only if they have the same elements.

### ZF2: Empty Set Axiom

$\exists x \forall y (y \notin x)$

There is a set with no elements.

### ZF3: Axiom of Pairs

$\forall x \forall \exists z \forall w (w \in z \iff w = x \vee w = y)$

Given any 2 sets (which may be the same set), there is a set which contains just those two.

## Axioms 4-6: Separation, Power Set, Union

### ZF4: Axiom of Separation**

$\forall x \exists y \forall z ( z \in y \iff (z \in x \wedge \phi(z) ))$, where $\phi(z)$ is any statement of the formal language with free variable z.

For any set x, there is a set y containing all elements z of x which meet the condition $\phi(z)$.

i.e. we can separate x into elements z which match the condition $\phi(z)$, and those which don't.

Note that the Axiom of Separation is not considered one axiom, but a whole series of axioms, one per formula. A term used is 


**We write $y = \{ x \in X: \phi(x) \}$ for a set y such that $\forall z( z \in y \iff ((z \in X) \wedge \phi(z)))$**


### ZF5: Power Set Axiom**

$\forall x \exists y \forall z (z \in y \iff z \subset x)$

For any set x, there is a set (called the power set of x), which contains all the subsets of x. We write:

**$P(X)$ for a set y such that $\forall z (z \in y \iff z \subset x)$**

### ZF6: Union Axiom**

$\forall x \exists y \forall z (z \in y \iff \exists w( z \in w \wedge w \in x))$

For any set x, there is a set which is the union of all its elements. We write:

$\bigcup X$ for a set y such that $\forall z (z \in y \iff \exists w( z \in w \wedge w \in x))$. 

Also, we write, for the union of any two sets:

$A \cup B = \bigcup \{ A, B \}$. 

Note that the Axiom of Pairs guarantees the set {A,B} exists.

Note: Expressing the Axiom as union of elements of a set allows us to allow arbitrary unions.

The union axiom allows us to extend our curly brace notation to more than 2 elements e.g. {x, y, z, ...}.

## Axioms 7-9: Infinity, Replacement, Foundation

### Axiom of Infinity

$\exists x(\emptyset \in x \wedge (y \in x \implies y \cup \{y\} \in x))$

There is an inductive set

### Axiom of Replacement

Given a "class function" $\phi(s,t)$, when its domain is restricted to x, the resulting images form a set y.

Consider $\phi(s,t):$ t = {s}. Given a set x, we can form a set $\{ \{ y \}: y \in x \}$. Though $\phi(s,t)$ looks like a function, it is not, but it can behave like one, once we have replaced s with values from some set x.


Note that the Axiom of replacement is not considered one axiom, but a whole series of axioms, one per formula. A term used is 

### Axiom of Foundation

$\forall x \exists y (y \in x \wedge x \cap y = \emptyset) $

Every set is well founded i.e. it contains a $\in$-minimal element.

This prevents forming sets e.g $a \in a$, or $b \in a \wedge b \in a$. 

For example, if we see N, it is composed of an infinite number of natural numbers, each of one contains the other. But starting from N, if we take any n in N, and then every m in n, and so on, eventually you will reach the empty set, in a finite number of steps. Thus a well-founded set has no infinite $\in$-descending chain.

This plays a key role in defining the cumulative hierarchy of sets, discussed under ordinals.



## Theorem 4.3 No infinite descending $\in$-chains

Suppose $\{x_n\}$ is a sequence. Then this sequence cannot form an infinite descending $\in$-chains.

## Proof

## The Recursion Principle

Let $y_0$ be an element of a set Y and assume a function $h:N \times Y \to Y$ exists. Then there exists a unique function $f:N \to Y$ such that:

$f(0) = y_0$

$f(n^+) = h(n, f(n))$ for all $n \in N$

## Proof of the Recursion Principle

Start by defining an n-function. For any n in N, $g:n^+ \to Y$, is an n-function if:

1) $g(0) = y_0$
2) $g(i^+) = h(i, g(i))$ for all $i \in n$

**Lemma**

Suppose g is an n-function and g' is an n'-function. Let m = min(n,n'). Then $g|m^+ = g'|m^+$

Proof: For 0: $g(0) = g'(0) = y_0$
Assume true for k < m, so g(k) = g'(k)

Then, $g(k^+) = h(k, g(k)) = h(k, g'(k)) = g'(k)$

**g_n exists for all n in N**

Let $A = \{n \in N: g_n \text{ exists } \}$

For 0, we have $g_0: 0^+ -> Y: 0 \mapsto y_0$

Assume $g_n$ exists for some n, i.e. n in A.

Then, define $g_{n^+} = \{(n^+, h(n, g_n(n)))\} \cup g_n$

Clearly $g_{n^+}$ has domain $n^{++}$. It is "single-valued" i.e. it is a valid function, since $g_n$ is a function only defined from 0..n, and we have added only one new value. For 0..n, g(i) = h(i,g(i)). For $n^+$ also we have defined this correctly. So $n^+$ is in A, and $g_n$ is defined for all n.

**$\bigcup \{g_n\}$ is a set**

If m < n, $g_m = g_m{\big|}_{m^+} = g_n{\big|}_{m^+} \subset g_n$

Thus the $g_n$'s form an $\subset$-chain. 

We want $f = \bigcup\{ g_n: n \in N\}$. However, to form the set, we need the axiom of replacement.

$\phi(s,t) = (s \in N \implies \text{ t is an s-function } \wedge (s \notin n \implies t = \emptyset ) )$ - the last part ensuring that the function is well defined for s not in N.

We have shown that for each n, $g_n is an n-function$, so that $\phi(s,t)$ is a class function. Taking x in the axiom of replacement to be N, we get the set of images as $\{g_n: n \in N}\$, which we are assured is a set.

**Forming f**

$f = \bigcup\{ g_n: n \in N\}$

It follows from ex 6.31 that f is a function, with domain equal to the union of the domains of g, i.e. N. Whenever m < n, we have $g_m \subset g_n \subset f$, so 
$f(m) = g_n(m)$.

Thus, $f(0) = g_n(0) = y_0$ for any n.

For any $n \in N$:

$f(n^+) = g_{n^+}(n^+) = h(n, g^{n+}(n)) = h(n, f(n))$

Thus f meets the required function criteria.

**Uniqueness of f**

Suppose f' is such that $f' \ne f$, but it meets the criteria. There must be a least natural number $m^+$, such that $f(m^+) \ne f'(m^+)$ - by the well-orderedness of N. Note that m cannot be zero, since $f(0) = f'(0) = y_0$

However, $f'(m^+) = h(m, f'(m))$ and $f(m^+) = h(m, f(m))$. Thus $f'(m) \ne f(m)$ which contradicts our assumption that $m^+$ is the least natural number such that $f(m^+) \ne f'(m^+)$.



## The Recursion Principle for N (Class form)

More generally than for the recursion principle, we would like to be able to define sets of the type $\{N, P(N), P(P(N), ...\}$ which are needed for Cantor's theory. We may define h(x,y) = P(y) (here $x \in N$), and define f as :

f(0) = N

$f(n^+) = P(f(n))$ for all n in N.

Unfortunately, this requires y to be a class with subset : $\{N, P(N), P(P(N)), \dots \}$, which is what we are trying to define.

Instead, here is a modified form of the recursion principle:

**Recursion Principle for N (class form)**

Given $\phi((x,y), z)$ a.k.a $\phi(x,y,z)$ be a formula such that for each $x \in N$ and set y, there is a unique set z such that $\phi(x,y,z)$ holds.

Thus, $\phi(x,y,z)$ essentially defines a class function $h:N \times U \to U$, where U is the universe. Let $y_0$ be any given set. Then there is a unique function f with domain N, such that:

$f(0) = y_0$

$f(n^+) = h(n, f(n))$, for all n in N.

We can then use the image of this function as a set, which would effectively be a sequence of sets $\{A_n: n \in N \}$. In fact this axiom is often used simply by assuming such a sequence of sets exists.

**If $X \subset Y$, then $T(X) \subset T(Y)$**

Let $f_X$, and $f_Y$ be the functions underlying the transitive closure of each set. We will show that $f_X(n) \subset f_Y(n)$ for each n.

a) $f_X(0) = X, f_Y(0) = Y$. Hence $f_X(0) \subset f_Y(0)$

b) Suppose $f_X(n) \subset f_Y(n)$, for some n in N. Then $f_X(n+1) = \bigcup f_X(n)$ and $f_Y(n+1) = \bigcup f_Y(n)$. i.e. 

$x \in f_X(n+1) \implies \exists x'(x' \in f_X(n) \wedge (x \in x'))$. But $f_X(n) \subset f_Y(n)$, by inductive step. Hence $x' \in f_Y(n) \implies x \in f_Y(n+1)$. Thus $f_X(n+1) \subset f_Y(n+1)$.

Thus, for any n, $f_X(n) \subset f_Y(n)$.

c) If x is in T(X), then $x \in f_X(n)$ for some n, hence $x in f_Y(n)$, hence $x \in T(Y)$.

**Let Z be such that: $Y \subset Z$, and whenever $x \in z \in Z$, then $x \in Z$** 

Informally Z is a $\in$-transitive set. Show that $T(Y) \subset Z$, i.e. $T(Y)$ is the smallest transitive set containing Y as a subset.

Is f(0) = Y subset of Z? yes, as given.

Suppose f(n) is in Z. Then if $z \in f(n)$, then any x in z is also in Z, by the definition of Z. Thus $\bigcup f(n) \subset f$ i.e. $f(n+1) \subset Z$. 

Based on this we conclude $T(Y) = \bigcup range(f) \subset Z$. If any $\in$-transitive contains Y, then $T(Y) \subset Z$ i.e. T(Y) is the smallest transitive set containing Y.

**By axiom of foundation, there is a $y \in T(Y)$ such that $T(Y) \cap y = \emptyset$**

Suppose $T(Y) \cap y = \emptyset$, $y \in T(Y)$. If y is not empty, then let $x \in y$ => $x \in T(Y)$. But then $x \in T(Y) \cap y$, which is not possible.

Hence y is the emptyset.
