Principle of Induction
If S is a subset of N, the set of natural numbers, such that :

i) 1 is an element in S
ii) If k belongs to S, so does k+1

Then S = N.

Proof By Induction
If P(n) is a statement indexed by N, with the following properties :

i) P(1) is true
ii) P(n) is true, then P(n+1) is true

then P(n) is true for all n in N.

This is equivalent to showing a set S = {n : P(n) is true} = N.
