## Topology based on Closure

**Theorem: Let X be a non-empty set. Let there be a collection of subsets which is closed under arbitrary intersections and finite unions. Then the complement of these sets form a topology on X, whose closed sets are precisely those given.**

Proof: Straightforward.

**Theorem: Let X be a non-empty set. If we define a closure operation which assigns to each set A of X a set $\bar A$ of X such that :  
a) $\bar{\emptyset} = \emptyset$  
b) $A \subseteq \bar A$  
c) $\bar{\bar A} = \bar A$  
d) $\overline{A \cup B} = \bar A \cup \bar B$**  

The 4 conditions are called the Kuratowski closure axioms.

If a closed set A is defined as one for which $A = \bar A$, then the complement of these sets form a topology whose closure operation is precisely that initially given.

Proof: 
By (a) empty set is closed.  
By (b) and (3), X is closed.
By (d) finite union of closed sets is closed.

(To show arbitrary intersection of closed sets is closed)

Given a collection $\{A_i\}$ of sets such that $A_i = \bar A_i$, we have to prove that $\cap_i A_i = \overline{\cap_i A_i}$. We already know from (b) that $\cap_i A_i \subseteq \overline{\cap_i A_i}$. So we just need to show $\overline{\cap_i A_i} \subseteq \cap_i A_i$.

We show that if $A \subseteq B$, then $\bar A \subseteq \bar B$. Given $A \subseteq B$, it follows that $B = A \cup B$ and by (d) $\bar B = \bar A \cup \bar B \implies \bar A \subseteq \bar B$. 

Thus, since $\cap_i A_i \subseteq A_i$ for each i, it follows that $\cap_i A_i \subseteq \bar A_i = A_i$ for each i. It follows that $\overline{\cap_i A_i} \subseteq \cap_i A_i$.

($\bar A$ is the smallest closed set containing A)

By (b) and (c), $\bar A$ is closed superset of A. We just have to show it is the smallest one.
For this, we have to show that if $A \subset B$ and $B = \bar B$, then $\bar A \subseteq B$. 

Since $A \subset B \implies B = A \cup B = \bar B = \bar A \cup \bar B$. Hence $\bar A \subseteq \bar B = B$.


## Open Bases and Open Subbases


**Definition (Basis)**: A given collection B of subsets of a set X is a basis of a topology in X if 
(1) The union of all sets in B is X  
(2) If x belongs to the intersection of two basis sets B1 and B2, then there is a basis element B3 containing x which is included in their intersection.  

**Theorem(Topology Induced by Basis) : The topology T generated by a basis B on a set X consists of all sets formed by the union of sets of B.**   
Proof: Clearly X is open as is the empty set (union of all, vs union of none). The union of any sets in B is an open set by definition. And consider a finite intersection of elements of B - any point in the intersection belongs to an open set entirely in the intersection. The union of these sets is the intersection itself. Thus T is a topology.

Thus, a set X is open in T if : (i) it is the union of sets of B or alternately (ii) given any x in X, there is a basis set contained in X which contains x (because this implies the X is a union of sets of B).

A **sub-basis** of a topology on set X is a collection of subsets of X whose union is X, and which generates a topology formed by the union of all finite intersections of the subsets.

A topology is **finer** than another, if it contains the other. A topology is **coarser** than another, if it contains the other.


## Countable Spaces

A **first countable space** is a topological space with a countable basis at each point.  
A **second countable space** is a topological space which has a countable basis.

**Theorem (Lindel&ouml;f's Theorem) : Let X be a second countable space. If a non-empty open set G in X is represented as the union of the collection $\{G_i\}$ of open sets, then G can be represented as a countable union of $G_i$'s.**  

Proof:  

If $\{B_n\}$ is a countable base, then for each $x \in G$, there is an i such that $x \in G_i$, and we can find a basic open set $B_n$ such that $x \in B_n \subseteq G_i$. The union of these $B_n$'s across all x in G is necessarily equal to G, and is necessarily countable (since $B_n$ is countable). Choosing the corresponding containing $G_i$ for each $B_n$, we get a countable collection of $G_i$'s which cover all of G.

**Theorem: Let X be a second countable space. Then any open base for X has a countable subclass which is also an open base.**  

Proof:

Let $G_i$ be an open base. As X is second countable, let $\{B_n\}$ be a countable base, then by the previous theorem, then every $B_n$ can be represented by a countable subcollection of $G_i$. And since $\{B_n\}$ is a countable base, we can form a countable subcollection of countable $\{G_i\}$'s which form a countable basis for X. 

**Theorem: Every separable metric space is second countable.**  

Let X be separable, and assume A is a countable dense subset of X. If we consider the open balls with rational radii centered on all the points of A, we then get a countable collection W of open sets. We show this is an open base.

Let G be an arbitrary non-empty open set. Let x be a point in G, then there is a ball $B_r(x)$ contained entirely in G. Also $B_{r/3}(x)$ is entirely in G. Now there must be an $a \in A$ in $B_{r/3}(x)$. And choosing an $r_1$ such as $r/3 < w < 2r/3$ there is a $B_w(a)$ contains x and is entirely in $B_r(x)$ and hence in G. But $B_w(a) \in W$, hence it follows that G is a union of a sets from W, a countable collection of open sets.




Example: Since open intervals form a base for R, intervals of the form $(a, +\infty)$ and $(-\infty,b)$ form a sub-basis of R.

**Theorem: Let X be a non-empty set, and S be an arbitrary collection of subsets of X. Then S can serve as a sub-basis for a topology on X, in the sense that the collection of all unions of finite intersections of sets in S is a topology.**  

Proof:  

(S is empty)  
If S is empty, then the intersection is X. And all unions of $\{X\}$ is $\{\emptyset, X\}$. This is a topology where every point is in the same neighborhood as every other.

(S is non-empty)  
Let B be the set of all finite intersections of sets in S - note that X is in this. Let T be the set of arbitrary unions of sets from S - so this includes $\emptyset$. Also T contains X, and T is closed under arbitrary unions. We need to prove T is closed under finite intersections.

Let $\{G_1, G_2, ..., G_n\}$ for $n > 0$ be a collection of sets from S. Let $G = \cap_{i=1}^n G_i$. If G is empty then it is in T. So assume G is not empty. Let x be a point in G, i.e. $x \in G_i$ for each i, and hence, by definition of T, for each $G_i$, there is a $B_i$, such that $x \in B_i \subseteq G_i$. The union of these $B_i$'s is equal to G. But each $B_i$ is the intersection of a finite number of sets from S. The intersection of these n $B_i$'s is 
a) itself an intersection of a finite numbers of sets of B and hence belongs to B
b) contains x
c) is a subset of G

Thus G is a union of sets of B, and is therefore in T.

T is the *topology* generated by the collection S.

## Compact Spaces

**Definition: Compact Space** A compact space is a topological space where every open cover has a finite subcover. A compact subspace of a topological space is a subspace which is compact as a topological space in its own right.

**Theorem: Any closed subspace of a compact space is compact.**

Proof:  

Let Y be a closed subspace of a compact space X. Considering Y as a subspace, the open sets in Y are the intersection of open sets in X with Y. Let $\{G_i\}$ be an open cover of Y in Y. Let $\{H_i\}$ be open sets in X such that $Y \cap H_i = G_i$. Then $Y^c \cup \{H_i\}$ form an open cover of X. Hence there is a finite subcover of X containing a finite number of $H_i$'s and possibly, $Y^c$, since X is compact. The finite $H_i$'s cover Y, and hence so do the corresponding finite number of $G_i$'s. These form a finite subcover of Y.

**Theorem: Any continuous image of a compact space is compact.**  

Proof:  
Let $f:X \to Y$ be a continuous, onto function, with X compact.  Let $G_i$ be an open cover of Y. Then $H_i = f^{-1}(G_i)$ forms an open cover of X, and hence has a finite subcover. Note that every x in X is in one of these (finite) $H_i$. And for these $H_i$, $f(H_i) \subseteq G_i$ for a finite number of $G_i$'s. Since f is onto, these finite $G_i$'s form a finite subcover. Thus Y is compact.

What if f is not onto? Then the argument above can be made for the subset f(X) of Y, taking f(X) as a subspace in its own right, and the open sets of f(X) as the open sets of Y which intersect with f(X). The rest follows.

**Theorem: A topological space is compact iff every collection of closed sets with empty intersection has a finite subcollection with empty intersection.**  

Proof:  
The union of the complements of these closed sets are open and cover the entire space i.e. they are an open cover, and hence have a finite subcover. Taking complements again, we get a finite subcollection of closed sets which has an empty intersection.

**Theorem: A topological space is compact iff every collection of closed sets with the finite intersection property has non-empty intersection.**   

Proof:  

(=>)  
Assume the space is compact.
In effect we are saying that no finite subcollection of these closed sets has an empty intersection. It follows from the previous theorem, that the union of the complements of the collection cannot be an open cover, since it could have no finite subcover. Thus the intersection must be non-empty.

(<=)  
Assume every collection of closed sets with the finite intersection property has non-empty intersection. Take an opem cover of the space. The complements are closed; their intersection is empty. Thus this collection of closed sets does not have fip - there is at least one finite subcollection with an empty intersection. The complements are open sets forming a finite subcover i.e the space is compact.

**Theorem: A topological space is compact if every basic open cover has a finite subcover**  

Proof:  Straightforward.

**Theorem: A topological space is compact if every subbasic open cover has a finite subcover, or equivalently, if every collection of subbasic closed sets with the finite intersection property has non-empty intersection.**  

Proof:  

We use closed basis / closed subbasis as a complement of open basis / open sub-basis. Consider a closed subbasis S and let $\{B_i\}$ is its generated closed basis i.e. the collection of all finite unions of the closed basis.

Let $\{B_j\}$ be a non-empty collection of $B_i$'s with a finite intersection property - we have to show that $\cap_j B_j$ is non-empty. We use Zorn's lemma to show that $\{B_j\}$ is contained in some maximal subcollection $\{B_k\}$ of $B_i$'s such that any collection of $B_i$'s containing $\{B_k\}$ does not have FIP.

($\{B_j\}$ is contained in a maximal subcollection $\{B_k\}$)  
Consider the family of all subcollections in $B_i$'s containing $\{B_k\}$, partially ordered by inclusion. Consider any chain in this family. The union of this chain has the FIP, because any finite collection of sets in the union belong to some member of the chain, and hence has a non-empty intersection. (consider any 2 points x and y in the union. Suppose x is in $B_1$ and y is in $B_2$. Since these are in a chain, either $B_1 \subseteq B_2$ or vice versa. So at least one contains both x and y. We can then do this repeatedly for any finite collection). Thus every chain in this family has an upper bound, which implies it has a maximal element $\{B_k\}$. 

Since $\cap_k B_k \subseteq \cap_j B_j$, we have to show $\cap_k B_k$ is non-empty. Each $B_k$ is a finite union of sets in our base e.g $B_1 = S_1 \cup S_2 \cup ... \cup S_n$. 

(At least one of $S_1, S_2, S_n$ belong to $\{B_k\}$)  
If $S_1$ does not belong to $B_k$, then $\{B_k, S_1\}$ does not have the FIP so $S_1$ is disjoint from some finite subcollection of $\{B_k\}$'s. If we take this for all $S_i$, then $B_1$ itself is disjoint from some finite subcollection of $\{B_k\}$'s and this would mean that if we add $B_1$ to the mix, we get a finite subcollection with an empty intersection which is not possible. For for $B_1$, at one of the $S_i$ belongs to $B_k$.

($\{B_j\}$ has a non-empty intersection)  
So every member of $\{B_k\}$ contains a subset which is also a member of $B_k$. The resulting subcollection of subbasis closed sets has the FIP, as they are a part of $\{B_k\}$, and hence has a non-empty intersection by the hypothesis. In this case $\cap_k B_k$ is non-empty, since every set in $\{B_k\}$ contains at least one of the subbasis closed sets. 

(Space is compact)  
It follows that every non-empty collection of closed basis sets with a finite intersection property has a non-empty intersection, and hence the complement i.e. every open basis cover has a finite subcover ie. the space is compact.



**Theorem (Heine-Borel Theorem) : Every closed and bounded subspace of the real line is compact.**  

A closed and bounded subspace of the real line is a closed subspace of some closed interval $[a,b]$. If a = b, this is clear, so we may assume that a < b. The intervals of the form $[a,d)$ and $(c,b]$ where $a < c < d$ and $a < d < b$ and is an open subbase.

Then, the set of all $[a,c]$'s and $[d,b]$'s is a closed subbase. 

Let $S = \{[a,c_i],[d_j,b]\}$ be a set of subbasic closed sets with the finite intersection property. We have to show the $\cap S$ is not empty. Note that if S contains only intervals of one type, then clearly the intervals all contain either a or b, hence the intersection is not empty.

Consider the case where S has intervals of both types : The intersection is non-empty only if $d = sup d_j \le c_i$ for all i's. This is easy to prove : Assume there is a $c_k < d$, this implies, there is a $d_k$, such that $c_k < d_k$. But then $[a,c_k] \cup [d_k,b]$ us empty violating the FIP property. So S has a non-empty intersection.

Thus the interval $[a,b]$ is compact. Hence, any closed subset is also compact.
