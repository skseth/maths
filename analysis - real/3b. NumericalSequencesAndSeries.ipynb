## Series

Given a name $\{a_n\}$, we use the notation :

$$\sum_{n=p}^{q} a_n = a_p + ... + a_q$$

With $\{a_n\}$, we then associate a sequence of $\{s_n\}$ :

$$\text{partial sum} = s_n = \sum_{k=1}^{n} a_k$$

An **infinite series** is denoted as :

$$\text{infinite series} = \sum_{n=1}^{\infty} a_n = \lim_{n \to \infty} s_n$$

It follows that the limit may or may not exist. If $\{s_n\}$ **converges**, we say :

$$\sum_{n=1}^{\infty} a_n = s \text{ ... where s is called the sum of the series, but is actually a limit}$$



## Convergence of Series

**Theorem (Cauchy Criterion for Series): $\sum a_n$ converges if and only if for every $\epsilon > 0$ there is an integer N such that, if $m \le n \le N$**

$$\left|\sum_{k=n}^{m} a_k\right| \le \epsilon$$

**In particular, by taking m = n, becomes : $|a_n| \le \epsilon$ , ($n \ge N$)**

Proof: 

The sequence $\{s_n\}$ converges (by the Cauchy Criterion) iff there is an P, such that :

$$d(s_m,s_p) = \left|\sum_{k=p+1}^{m} a_k\right| \le \epsilon$$  

For all $m,p \ge P, m > p$. We can choose n = p + 1, and N = P + 1

**Theorem: If $\sum a_n$ converges, then $\lim_{n \to \infty} a_n = 0$**

Proof: Take m = n in the theorem above. We get $|a_n| \le \epsilon$, whenever $n \ge N$.

Note: Converse is not true : $\lim_{n \to \infty} a_n = 0$ does not imply $\sum a_n$ converges. For example $\sum 1/n$ diverges.

**Theorem (Monotonic Convergence for Series): A series of nonnegative terms converges if and only if its partial sums form a bounded sequence**

Proof: If all terms are "nonnegative", then $\{s_n\}$ is monotonically increasing. It follows that $\{s_n\}$ converges iff $\{s_n\}$ is bounded.


**Theorem (Comparison Test for Series)**  
**a) If $|a_n| \le c_n$ for $n \ge N_0$, where $N_0$ is some fixed integer, and if $\sum c_n$ converges, then $\sum a_n$ converges.**  
**b) If $a_n \ge d_n \ge 0$ for $n \ge N_0$, and if $\sum d_n$ diverges, then $\sum a_n$ diverges. Note that both $a_n$ and $d_n$ are non-negative.**

Proof:

$$\left|\sum_{k=n}^{m} a_k\right| \le \sum_{k=n}^{m} |a_k| \le \sum_{k=n}^{m} c_k $$

Since $\sum c_n$ converges, by Cauchy criterion the result follows.

**Theorem (Geometric Series Convergence): If $0 \le x < 1$, then $\sum_{n = 0}^{\infty}x^n = \frac{1}{1 - x}.$**
**If $x \ge 1$, the series diverges.**

If $x \ne 1$,

$$s_n = \sum_{k = 0}^{n} x^k = \frac{1 - x^{n+1}}{1 - x}$$

As $n \to \infty$, $x^{n+1} \to 0$. 

**Theorem (Nonnegative Monotonically Decreasing Terms) : Then the series $\sum_{n=1}^{\infty} a_n$ converges if and only if the series converges.**

$$\sum_{}^{} 2^ka_{2^k} = a_1 + 2a_2 + 4a_4 + 8a_8 + ...$$

Proof:

The partial sums :

$$s_n = a_1 + a_2 + ... + a_n$$
$$t_k = a_1 + 2a_2 + ... + 2^ka_{2^k}$$

For $n < 2^k$

$$
\begin{align*}
s_n &\le a_1 + (a_2 + a_3) + ... + (a_{2^k} + ... + a_{2^{k+1}-1})  \\
&\le a_1 + 2a_2 + + 2^ka_{2^k} \text{  monotonically decreasing terms}  \\
&= t_k  \\
\end{align*}
$$

For $n > 2^k$

$$
\begin{align*}
s_n &\ge a_1 + a_2 + (a_3 + a_4) + ... + (a_{2^{k-1}+1} + ... + a_{2^k}) \\
&\ge (1/2)a_1 + a_2 + 2a_4 + 2^{k-1}a_{2^k} \\
&= \frac{1}{2}t_k
\end{align*}
$$


Given the two results, these sequences are either both bounded or both unbounded.

**Theorem : $\sum \frac{1}{n^p}$ converges if p > 1 and diverges if $p \le 1$** 

If $p \le 0$, then the sequence $\{1/n^p\}$ diverges => the series diverges.

If p > 0, then the terms $\{1/n^p\}$ are nonnegative & monotonically decreasing. So consider the following series :

$$\sum 2^k \frac{1}{(2^k)^p} = \sum 2^{k(1 - p)}$$

If $p > 1$, $2^{1-p} < 1$. The RHS then is a convergent geometric series i.e. this series converges and by the previous theorem $\sum \frac{1}{n^p}$ converges.

**Theorem: If p > 1,
$$\sum_{n=2}^{\infty} \frac{1}{n(\log{n})^p}$$
converges; if $p \le 1$, the series diverges.**  

$\log{n}$ is monotonically increasing, $\frac{1}{n(\log{n})^p}$ decreases. Consider the following :

$$\sum 2^k \frac{1}{2^k(\log{2^k})^p} = \frac{1}{(\log{2})^p} \sum \frac{1}{k^p}$$

The series above converges when p > 1, diverges when $p \le 1$. 

## The Number e

**Theorem (Definition of e): $$e = \sum_{n=0}^{\infty} \frac{1}{n!}$$**

Proof:
$$
\begin{align*}
s_n &= 1 + 1 + \frac{1}{2!} + ... + \frac{1}{n!}  \\
&< 2 + \frac{1}{2} + \frac{1}{2^2} + ... + \frac{1}{2^n} \\
&< 3
\end{align*}
$$

The series in bounded, and monotonically increasing. 


**Theorem (e as the Ultimate Compound Interest)**

$$e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^{n}$$

Let $$s_n = \sum_{k=0}^{n} \frac{1}{k!}$$

$$
\begin{align*}
t_n = \left(1 + \frac{1}{n}\right)^{n} &= 1 + n\frac{1}{n} + \frac{n(n-1)}{2!}\frac{1}{n^2} + ... + \frac{n(n-1)..(1)}{n!}\frac{1}{n^n} \\
&= 1 + 1 + \frac{1}{2!}(1-1/n) + ... + \frac{1}{n!}(1 - 1/n)(1 - 2/n).. + ...  \\
&\le s_n
\end{align*}
$$

Hence $t_n \le s_n$, so that 

A) $\limsup_{n \to \infty} \le \epsilon$

Next, if $n \ge m$,

$t_n \ge 1 + 1 + \frac{1}{2!}(1-1/n) + ... + \frac{1}{m!}(1 - 1/n)(1 - 2/n)..(1 - \frac{m-1}{n})$

Let $n \to \infty$
$\liminf_{n \to \infty} t_n \ge 1 + 1 + \frac{1}{2!} + ... + \frac{1}{m!} = s_m$

Let $m \to \infty$, 

B) $e \le \liminf_{n \to \infty} t_n$

From (A) and (B), we get $\lim_{n \to \infty} t_n = e$


**Rapidity of Convergence of e**


$$
\begin{align*}
e - s_n &= \frac{1}{(n+1)!} + \frac{1}{(n+2)!} + \frac{1}{(n+3)!} + ...  \\
&< \frac{1}{(n+1)!}\left(1 + \frac{1}{n+1} + \frac{1}{(n+1)^2} + ...\right) \\
&= \frac{1}{n!n}
\end{align*}
$$

So, $0 < e - s_n < \frac{1}{n!n}$




## The Root And Ratio Tests



**Theorem (Root Test) : Given $\sum a_n$, put $\alpha = \limsup_{n \to \infty} \sqrt[n]{a_n}$  
Then :  
(a) if $\alpha < 1$, $\sum a_n$ converges  
(b) if $\alpha > 1$, $\sum a_n$ diverges  
(c) if $\alpha = 1$, the test gives no information**

Proof:

**a. ($\alpha < 1$)**  
we can choose $\beta$ so that $\alpha < \beta < 1$, and an integer N, such that for $n \ge N$ $\sqrt[n]{|a_n|} < \beta$

Thus, for $n \ge N$, $|a_n| < \beta^n$

If $0 < \beta < 1$, $\sum \beta^n$ converges. By the Comparison Test, so does $\sum a^n$

**b. ($\alpha > 1$)**  

i.e. $\limsup_{n \to \infty} \sqrt[n]{a_n} > 1$. This implies there is an sequence $\{n^k\}$ such that :

$\sqrt[n^k]{|a_{n^k}|} \to \alpha$ - thus $|a_n| > 1$ for infinitely many values of n => $a_n \to 0$ does not hold, which implies $\sum a_n$ diverges.

**c. ($\alpha = 1$)**  

Consider $\sum 1/n$ and $\sum 1/n^2$. $\alpha = \limsup_{n \to \infty} \sqrt[n]{1/n} = 1$, the same for the other case. But  $\sum 1/n$ diverges, $\sum 1/n^2$ converges.

**Theorem (Ratio Test) : The series $\sum a_n$  
(a) converges if $\limsup_{n \to \infty} \left|\frac{a_{n+1}}{a_n}\right| < 1$  
(b) diverges if $\left|\frac{a_{n+1}}{a_n}\right| \ge 1$ for all $n > n_0$ where $n_0$ is a fixed integer**

**(a)**  
If condition (a) holds, we can find $\beta < 1$, and an integer N, such that $\left|\frac{a_{n+1}}{a_n}\right| < \beta$, for $n \ge N$.

Then $|a_{N+1}| < \beta|a_N| \implies |a_n| < \beta^{n-N}|a_N| = \beta^{-N}|a_N|\beta^n$  

It follows from the comparison test that since $\beta^n$ converges, so does $\sum a_n$.

**(b)**  
If $a_n + 1 \ge a_n$ for all $n \ge n_0$, then choose $0 < \epsilon < |a_{n_0}|$. For any $N > n_0$, $|a_n - 0| > \epsilon$


**Theorem (Root Test more Sensitive than Ratio Test)**

For any $\{c_n\}$ of positive numbers :

$$
\begin{align*}
\liminf_{n \to \infty} \frac{c_{n+1}}{c_n} &\le \liminf_{n \to \infty} \sqrt[n]{c_n}  \\
\limsup_{n \to \infty} \sqrt[n]{c_n} &\le \limsup_{n \to \infty} \frac{c_{n+1}}{c_n}
\end{align*}
$$

Proof:  
Let
$$\alpha = \limsup_{n \to \infty} \frac{c_{n+1}}{c_n}$$
$$\gamma = \liminf_{n \to \infty} \frac{c_{n+1}}{c_n}$$

($\alpha = +\infty, \gamma = -\infty$)  
Trivial

($\alpha$ finite, $\gamma$ finite)  
If $\alpha$ is finite, choose $\beta < \alpha$. There is an integer N such that for all $n \ge N$

$$\frac{c_{n+1}}{c_n} \le \beta \implies c_n \le c_N\beta^{-N}\beta^n \implies \sqrt[n]{c_n} \le \sqrt[n]{c_N\beta^{-N}}\beta$$

Then $$\limsup_{n \to \infty} \sqrt[n]{c_n} \le \beta$$

If $\gamma$ is finite, choose $\beta < \gamma$. There is an integer N such that for all $n \ge N$

$$\frac{c_{n+1}}{c_n} \ge \beta \implies c_n \ge c_N\beta^{-N}\beta^n \implies \sqrt[n]{c_n} \ge \sqrt[n]{c_N\beta^{-N}}\beta$$

Then $$\liminf_{n \to \infty} \sqrt[n]{c_n} \ge \beta$$

($\alpha = -\infty, \gamma = +\infty$)
