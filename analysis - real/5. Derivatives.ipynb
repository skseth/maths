5.1 Definition : f is defined and real-valued on [a,b]. For any x in [a,b] we define 

f’(x) = lim (t->x) f(t) - f(x) / t - x  

Note : limit of the slope of the secant line of the graph.

If f is continuous in [a,b], is f differentiable in [a,b]? No

If f is differentiable in [a,b] must f be continuous? lim t->x f(x) - f(t) = lim t->x [f(t) - f(x) / t - x] (t-x) = f’(x). 0 = 0. Hence f is continuous.

But f f is differentiable, f’ be continuous? No. But f’ always satisfy IVP and f’ has no simple discontinuity.

Example. f(x) = x4/3 Sin(1/x), = 0 at 0.

Call a function f a C1 - function if f’ exists and is continuous. Ck function if it has k derivatives and kth derivative is discontinuous. Powers of x can be used to create such sequences.

C0 is continuous functions. Cinf - if all derivatives exists. Smooth functions. 

See http://www.ugrad.math.ubc.ca/coursedoc/math100/notes/zoo/continuous.html

For C and Rk

For C, if f(t) = f1(t) + i f2(t), where f1 and f2 are real functions, then the same definition if applied shows that :

f’(x) = lim t->x f1(t) + i f2(t) - f1(x) - i f2(x) / t - x = f1’(x) + i f2’(x)

For Rk ,  let p(t) = f(t) - f(x) / t - x, where x, t are in Rk. Then, we define f’(x) as a point of Rk for which :

lim (t->x) | p(t) - f’(x) | = 0.

If f = (f1, f2, …, fk) then f’ = (f’1, f’2, …, f’k) and f is differentiable at x if and only if each of the component functions are differentiable at x.


Theorem 5.2 : Let f be defined on [a,b]. If f is differentiable at some x in [a,b], f is continuous at x.

Proof : Let f’(x) be defined. lim t-> x f(t) - f(x) = lim t->x f’(x). (t-x) = f’(x) lim t->x (t - x) = 0.

Definition (Local Maximum / Minimum) : Let f be a real function on a metric space X. We say f has a local maximum at point p in X, if there exists d > 0 such that f(q) <= f(p) for all q in x with d(p,q) < d.

Note : similar defn applies to local minimum.

5.8 Theorem (Local Maximum / Minimum) : Let f be defined on [a,b]. If f has a local maximum at some x in (a,b) and f’(x) exists, then f’(x) = 0.
Note : Similar statement is true for local minimum.

Proof : Since f’(x) exists, lim (f(x+h) - f(x)/ h  = lim f(x-h) - f(x) / -h = f’(x)

Since x is a local minimum, the first is a limit of negative terms, and the second a limit of positive terms - if the limits are same, their common limit f’(x) must be zero.

5.9 The Generalized Mean Value Theorem : 

If f and g are continuous real functions on [a,b] which are differentiable in (a,b), then there is a point x in (a,b) at which :

[f(b) - f(a)] g’(x) = [g(b) - g(a)] f’(x)

Note : Differentiability is not required at endpoints.

Proof : Let h(t) = [f(b) - f(a)]g(t) - [g(b) - g(a)] f(t)
h is continuous & differentiable in (a,b) and h(a) = f(b)g(a) - f(a)g(b) = h(b). 

h(x), being a continuous function on a compact set has a maximum and minimum and attains these bounds. Let t be a point in (a,b) where h(t) is maximum - then clearly, by theorem 5.8, h’(t) = 0. Similar argument applies for the minimum. 

Note that h(x) may be constant - this this case h’(x) = 0 for all x in (a,b). Also h(x) may have only a maximum, or only a minimum.


The Mean Value Theorem : If f is continuous on some closed interval [a,b], and differentiable on the open interval (a,b), then there is a point c in (a,b) such that :

f(b) - f(a) = (b-a) f’(c) i.e. derivative of some point in the interval has same slope as the line joining a and b.

Proof : Apply the Generalized Mean Value Theorem, with g(x) = x.

5.11 Theorem  : Suppose f is differentiable in (a,b). If for all x in (a,b)

i) f’(x) >= 0, f is montonically increasing
ii) f’(x) = 0, f is constant
iii) f’(x) <= 0, f is monotonically decreasing

Proof : 
Since f’(x) exists, lim (f(x+h) - f(x)/ h  = lim f(x-h) - f(x) / -h = f’(x)

(i) => f(x+h) > f(x), f(x-h) < f(x) for all x.
(ii) => f(x +h) = f(x) for all x
(iii) implies opp of (i)

5.12 Theorem (Intermediate Value Property of Derivatives) : Suppose f is a differentiable function on [a,b] and suppose f’(a) < q < f’(b). Then there is a point x in (a,b) such that f’(x) = q.

Proof : Put g(t) = f(t) - qt. Then g’(t) = f’(t) - q.

g’(a) = f’(a) - q < 0. g’(b) = f’(b) - q > 0.

Since g’(a) < 0, there is a point t1 in (a,b) such that g(t1) < g(a). Similarly, g’(b) > 0 => there is a t2 in (a,b) such that g(t2) < g(b). Hence, g, being a continuous function on a compact set [a,b], has a minimum, and it lies in (a,b). (Note : Rudin mentions this as [a,b] ). Let x be such a minimum. Then g’(x) = 0 i.e. f’(x) = q.

Corollary : If f is differentiable on [a,b], f’ cannot have any simple discontinuities.

Proof : If f’(x+) and f’(x-) exist, then for any interval around x, all values between f’(x+) and f’(x-) also exist i.e f’(x+) = f’(x-) = f’(x) for all x, if the limits exist. Hence no simple discontinuities. The only possible discontinuity is if limits don’t exist at a point e.g. x2 Sin(1/x) at 0.

Theorem 5.13 L’Hospital’s Rule : Suppose f and g are real and differentiable in (a,b) and g’(x) <> 0 for all x in (a,b), where a,b may be in the extended reals. Suppose 

f’(x) / g’(x) -> A as x -> a. 

If, in addition, f(x) -> 0 and g(x) -> 0 as x -> a, or if g(x) -> +inf. as x-> a, then : f(x) / g(x) -> A as x -> a.

The result is equally true for x -> b, or g(x) -> -inf.

Proof : First consider case where -inf <= A < infinity. Choose q so that A < q, and choose A < r < q. This covers the cases where A = -inf., or A is real. 

Now, f’(x) / g’(x) -> A as x -> a. Hence, there is a c in (a,b), such that a < t < c => f’(t) / g’(t) < r. 

Let a < x < y < c. There must be a point t between x and y (by the generalized MVT) s.t. :

f(x) - f(y) / g(x) - g(y) = f’(t) / g’(t) < r.... (1)

Keeping y fixed, consider x, as x -> a.

If we consider the case where f(x) and g(x) -> 0, as x -> a, then we see that f(y) / g(y) <= r < q, if we choose x close enough to a. 

Now, consider the case where g(x) -> inf. We can find a point c1 such that g(x) > g(y) and g(x) > 0 for a < x < c1. Rearranging (1) we get :

f(x) - f(y) < r (g(x) - g(y))  
i.e. f(x) / g(x) < r - r g(y) / g(x) + f(y) / g(x). for a < x < c1.

Letting x-> a, we can find a c2  in (a,c1) such that f(x) / g(x) < q for a < x < c2.

Hence we find that if A is real, or -infinity, there is an interval (a, c2) where f(x) / g(x) < q.

Similarly, if -inf < A <= +inf., choose p < A. We can then find a c3 such that given a < x < c3 , f(x) / g(x) > p.

Thus, for any real A, we p < f(x) / g(x) < q for a < x < min(c3, c2). Since p and q are arbitrary, we can get as close to to A as we want i.e. f(x) / g(x) -> A as x -> a.

If A is -inf. by the first result, we get f(x) / g(x) < q for any arbitrary q. And if A is +inf, we get f(x) / g(x) > p given any arbitrary p. 

Thus we see that lim f(x) / g(x) = lim f’(x)/g’(x) as x-> a whenever f(x) = g(x) -> 0 or g(x) -> inf. as x-> a. Similarly we can prove the result when g(x) -> -infinity, or x -> b.

5.15 (Taylor’s Theorem) : Let n be a positive integer, f be a real function defined on [a,b] - fn-1(t) is continuous on [a,b], fn(t) exists in (a,b). 

Let a’ and b’ be two points in [a,b].

Define P(t) = S(k=0 to n-1) (f(k)(a’) / k!) (t - a’)k where t is in [a,b].

Then there is a point x between a’ and b’ such that :

f(b’) = P(b’) + (f(n)(x) / n!) (b’ - a’)n

For n = 1 this is the mean value theorem.

Proof : Let us note that P(a’) = f(a’), P’(a’) = f’(a’), P(n-1) (a’) = f(n)(a’)

Let M be such that :

f(b’) = P(b’) + M(b’ - a’)n   … (1)

Define : g(t) = f(t) - P(t) - M (t - a’)n

g(n)(t) = f(n)(t) - n!M

Also g(a’) = g’(a’) = g’’(a’) = g(n-1)(a’) = 0  (since P(k)(a’) = f(k)(a’))

Also, g(b’) = 0 - by our choice of M. Hence we find, by MVT, an x1 between a’ and b’ such that g’(x1) = 0. We can then find x2 between a’ and x1, and, ultimately, an xn between a’ and xn-1 such that :

g(n)(xn) = 0 i.e. f(n)(xn) - n!M = 0 i.e. M = f(n)(xn) / n!. Hence

f(b’) = P(b’) + (f(n)(xn) / n!)(b’ - a’)n
