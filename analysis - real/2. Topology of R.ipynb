### Compactness

open cover  
A subset K of a metric space X is said to be compact if every open cover of K contains a finite subcover

Theorem: Suppose $K \subset Y \subset X$. K is compact relative to X, iff K is compact relative to Y. This makes it sensible to talk of a compact metric space.

Theorem: Compact subsets of metric spaces are closed.

Theorem: Closed subsets of compact sets are compact.

Corollary: If F is closed and K is compact, then $F \cap K$ is compact.

Theorem: If {Ka} is a collection of compact subsets of a metric space X such that the intersection of every finite subcollection of {Ka} is nonempty, then intersection of {Ka} is nonempty.

Theorem: If {Kn} is a sequence of nonempty compact sets such that Kn => Kn+1 (n = 1, 2, 3•...), then  Kn is not empty.

Corollary: If {Kn} is a nested sequence of nonempty compact sets, then the intersection is not empty

Theorem: If E is an infinite subset of a compact set K, then E has a limit point in K.

Theorem: If {In} is a sequence of nested intervals in R, the intersection of In is not empty

### Heine-Borel Theorem for R

Let $[a,b]$ be a closed, bounded interval of R. Every open cover of $[a,b]$ has a finite subcover i.e it is compact.



### Bolzano-Weierstrass Theorem

Bolzano-Weirstrass Theorem: Any bounded, infinite subset of R has an accumulation point.

**2. Prove that algebraic numbers are countable.**

Very basic description of proof for countability of algebraic numbers.

From http://www.math.brown.edu/~res/MFS/handout8.pdf :

To show that the set of algebraic numbers is countable, let Lk denote the set of algebraic
numbers that satisfy polynomials of the form c0 +c1x+...+cnxn where n < k and max(|cj|) < k. Note that there are at most kk polynomials of this form, and each one has at most k roots. Hence Lk is a ﬁnite set having at most kk+1 elements. As above, we make our list L1, L2, L3 of all algebraic numbers and weed out repeaters.


Basic Proof :

1. Consider 

All polynomials with coefficients a0, a1, a2, … an such that

let p(x) = anxn + an-1xn-1 + an-2xn-2 + … +  a0 . Each such polynomial has a root, and we can easily prove that it has no more than n roots by simple induction on n.

We define the height of a polynomial as 

h(p) = |a0| + |a1| + … + |an| + n 

Let S(k)  = { x : x is set of all solutions of polynomial with height less than k}.

Example : S(1) = {}, S(2) = {0}, S(3) = {0,1,-1}

What is the cardinality of S(N)? Clearly each of |a0|, |a1|, … <= N. 

For each n, the total combinations therefore are less than (2N + 1)n, where n <= N (including negative values and zero). Since n can be from 1 to N, we can say the total number of combinations (distinct equations) is less than N.(2N+1)N . Each equation will have no more than n roots, as can be easily proved using induction. Hence the number of roots (cardinality of S(N)) is no more than N2(2N + 1)N which is a finite number.

Since every integer i is a root of a polynomial (x = i), hence the algebraic numbers are infinite in number. 

Also {S(k)} form a countable sequence of finite sets containing the solutions to all polynomial equations. We can form a sequence {S’(k)} where each S’(k) = S(k) - U S(j), where 1 <= j <= k-1, to eliminate duplicates. Since each S(k) is finite, so is each S’(k), and the union of {S’(k)} is the set of all algebraic numbers which are there countable.

Thus every neighborhood of x has uncountable points of E, and x is in P. Thus p is closed. Also, since P is a set of all condensation points of E, the rest of E has at most countable points. Since each neighborhood of a point in P has uncountable points of E, but at most countable points of E/P, each point in P is a also a limit point of P. Hence P is perfect (though possibly empty).  Thus E is the union of a perfect set and at most countable other points.

Corollary : Every countable closed set E in Rk has isolated points.

If E had only limit points, E would be perfect and hence uncountable. 

29. Prove every open set in R is a union of at most countable collection of disjoint segments.

Let G be an open set in R.

For each x in G, define Ix = { y : (x,y) in G}. 
Ix is  open, since it is the arbitrary union of open sets. Also, since for any a, b, a < b in Ix, (x,a) and (x,b) are in Ix, and hence so is the intersection (a,b), Ix is connected.

Also, if x’ in G, x’ <> x, then either Ix’ = Ix or Ix’ n Ix = empty. Thus {Ix} forms a partition of G, i.e. 
G = U {Ix} is a union of disjoint segments. 
Since Q is dense in R, select one qx in each Ix . Since {qx} is at most countable, {Ix} is at most countable. Hence G is a union of at most countable collection of disjoint segments.
