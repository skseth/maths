# Topology 

## General

Inspired by the $\epsilon-\delta$ definition in classic continuity, but not wanting to restrict ourselves to metric spaces, we try to define a "neighborhood" of a point by mimicking the properties of open intervals, or open balls in R.

Let X be a set, and T be a collection of subsets of X. We say T is a **topology** on X if :

(1) X, $\emptyset$ are in T  
(2) Any union of a family of sets in T is also in T   
(3) A finite intersection of sets in T is also in T  

(X,T) is a **topological space**. A set in T is called an open set. 

In the rest of this section, X is a topological space.

Now, a set is called a **neighborhood** of a point x in X, if it contains an open set containing x. An open set is a neighborhood for each of its points.

A set is closed, if its complement is **open**. A set may be closed, open, both or neither. E.g. X and $\emptyset$ are closed and open. We immediately get that the intersection of any family of closed sets is closed, and the union of a finite family of closed sets is closed (apply de Morgan's Laws).

A point p is called a **limit point** of a subset A of X, if every neighborhood of p contains a point of A - {p} (p need not be in A, but we exclude it if it is).

**Theorem: A set is closed iff it contains all its limit points.**  

(=>) Let A be closed. Then X - A is open. X - A is a neighborhood of each of its points, so no point of X - A is a limit point of A.

(<=) Let A be a set containing all its limit points. Then consider $A^c$ - if A is $\emptyset$ or X, we are done. Otherwise every x in $A^c$ contains an neighborhood contained entirely in $A^c$, each of which contains an open set containing x. The union of these open sets is $A^c$ itself, and hence $A^c$ is open, and A is closed.

A **closure** of a set A is the set together with its limit points, and is called $\bar A$

**Theorem: The closure of a set A is the smallest closed set containing A.**  
First we prove $\bar A$ is closed. Let p be a limit point of $\bar A$, but in $X-\bar A$. Then there is a neigborhood of p containing no point of A (since p is not a limit point of A). But then each open neigborhood contains a limit point of A - but by definition it is also a neighborhood to the limit point, and hence contains points of A, a contradiction. Hence $\bar A$ is closed.

Let A be contained in A', a closed set. Then any limit point of A is a limit point of A', and hence in A'. Thus A' contains $\bar A$

**Corollary: A set is closed, iff it is equal to its closure**  

A set whose closure is the whole set is said to be **dense** in the set.

The **interior** of a set A is the union of all open sets contained in A. A point p lies in the interior of A iff A is a neighborhood of p. An open set is its own interior.

The **boundary** or **frontier** of a set A is the set of all points which are in the intersection of the closure of A and X-A. Think of the boundary of a circle for example. An alternative definition is the set of all points not in the interior of X or X-A. Proof: Any point in the frontier is a limit point of both A and X-A. Thus such points have every open neighborhood containing points of both A and X-A, and are therefore not in the interior of either.


**Definition (Basis)**: A given collection B of subsets of a set X is a basis of a topology in X if 
(1) The union of all sets in B is X  
(2) If x belongs to the intersection of two basis sets B1 and B2, then there is a basis element B3 containing x which is included in their intersection.  

**Theorem(Topology Induced by Basis) : The topology T generated by a basis B on a set X consists of all sets formed by the union of sets of B.**   
Proof: Clearly X is open as is the empty set (union of all, vs union of none). The union of any sets in B is an open set by definition. And consider a finite intersection of elements of B - any point in the intersection belongs to an open set entirely in the intersection. The union of these sets is the intersection itself. Thus T is a topology.

Thus, a set X is open in T if : (i) it is the union of sets of B or alternately (ii) given any x in X, there is a basis set contained in X which contains x (because this implies the X is a union of sets of B).

A **sub-basis** of a topology on set X is a collection of subsets of X whose union is X, and which generates a topology formed by the union of all finite intersections of the subsets.

A topology is **finer** than another, if it contains the other. A topology is **coarser** than another, if it contains the other.


## Continuity

A function is said to be **continuous** iff the pre-image of any open set (in the range) is open (in the domain).

[Stackexchange - Why formulate in terms of inverse images or pre-image](https://math.stackexchange.com/questions/1379363/why-formulate-continuity-in-terms-of-pre-images-instead-of-image) : Take f(x) = sin(x) on the inteval (-10,10) - the image is \[-1,1\] - i.e. closed. Same thing can be seen with closed sets. Compact sets map to compact sets, but sometimes that happens with discontinuous functions also, so its not an iff condition. 

Functions are many-one in general (and one-one sometimes). Inverse images are basically one-many. If you go from (say) an open set in the domain to an open set in the range, you are not getting the full picture - you are missing the other sets in the domain which may also be mapping to the same set in the range. So the inverse image allows you to characterize the entire behaviour for at least that part of the range. There is no way to take a set in the domain to fully characterize the behaviour of the range.

**Theorem: The composition of two continuous functions is a continuous function.**  
Proof: Let f: X -> Y, and g: Y -> Z. Let O be an open set in Z. Note that $(g\; o\; f)^{-1}(O) = f^{-1}\; o\; g^{-1}(O)$. Now, since g is continuous, $g^{-1}(O)$ is open in Y, and because f is continuous $f^{-1}\; o\; g^{-1}(O)$ is open in X.

**Theorem: A function is continuous iff the pre-image of an closed set is closed.**  
Proof: Assume f:X -> Y
(=>) Assume f:X -> Y is such that for any closed set C in Y, the pre-image $f^{-1}(C)$ is closed. Given any open set O in Y, $O^c$ is closed, and hence $f^{-1}(O^c)$ is closed => $(f^{-1}(O))^c$ is closed, since the inverse image of a complement is a complement of the inverse. Thus $(f^{-1}(O))$ is open.

(<=): A function is continous. Let O be an open set in Y. then $f^{-1}(O)$ is open. Given a closed set C in Y, $f^{-1}(C^c)$ is open, hence $(f^{-1}(C))^c$ is open. Hence $f^{-1}(C)$ is closed.

**Theorem: A function is continuous iff the image of a closure is included in the closure of the image.**  

e.g. If f:R -> R is a continuous function , then then $f([a,b]) \subset \overline{f((a,b))}$

If f:X -> Y, and A is a subset of X, then $f(\overline A) \subset \overline{f(A)}$.  
Proof: 
(=>) Let C be any closed set in Y. Then we have to show $D = f^{-1}(C)$ is closed in X. $f(\overline D) \subset f(\bar D) = C$. But this means $f^{-1}(\overline D) \subset f^{-1}(C) = D.$. Closure of D is included in D i.e. D is closed.

(<=) Let f be continuous. Then A be any set in X. Let B = f(A). Then $f^{-1}(\overline B) \supset A$ is closed, since f is continuous and of course all closures are closed. Thus, $f^{-1}(\overline B) \supset \overline A => f(\overline A) \subset \overline{f(A)}$

**Theorem: A function is continuous iff the closure of an inverse image of any set in the domain contains the closure of that set.**  



**Theorem: A function is continuous iff the inverse image of every basic open set in the range is open in the domain.**  

**Theorem: A function is continuous iff inverse image of neighborhood of any point in the image, is mapped to a neighborhood of the point.**
(=>) Assume f:X -> Y is continuous. Let V be a neighborhood of f(x). Then V contains an open set O containing f(x), and $f^{-}(O) \subset f^{-1}(N)$ is open and contains x i.e. $f^{-1}(N)$ is also a neigborhood of x.


A **homeomorphism** is a one-one, onto, continuous function with a continuous inverse between two topological spaces. A property of a space that is preserved by a homeomorphism is called a topological property. We will see such properties in subsequent discussions.

Some implications of homeomorphisms are surprising. For example, (0,1) and the real number line are homeomorphic, while \[0,1\] is not homeomorphic. Intuitively, this tells us that (0,1) and real number line have the same "shape" in some important sense (no gaps, fuzzy ends). We can conclude that "boundedness" is not a topological property. If we attach +- infinity to R, then suddenly it becomes homeomorphic to \[0,1\].


## Derived Topologies


**Theorem (Subspace Topology)**: Given a topological space (X,T) and any subset A, we can define a topology $T_A$ such that $T_A = \{ A \cup O : O \in T\}$. Then (A,$T_A$) is a **topological subspace** of X, and $T_A$ is the **induced topology** of the subspace A from the containing space X.  
Proof (that A is a topology) : Since A is a subset of X, $X \cap A = A$ is in $T_A$, since X is in T, and $\emptyset \cap A = \emptyset$ is in $T_A$. The union of any open sets in A and the intersection of a finite set of open sets in A can similarly be shown to be in $T_A$.

Note : The basis of this topology is the intersection of A with the basis sets. In general open sets in A are not open in X, but if A is open, then every open set in A is also open in X, as immediately follows from the definitions.

**Theorem (Order Topology)**: Given a totally ordered set X,with order relation <, the order topology is formed by a basis of all open intervals of X, and half-open intervals containing the largest / smallest element if any. Note that in R, the order topology is just the open intervals, since there is no largest / smallest element. 

An **interval** between points a and b could be (a,b), \[a,b), (a,b\], \[a,b\], a < b. The half-open intervals are called **rays** if the include the largest / smallest element.

In an order topology, the supremum of a set, if it exists, always belongs to its closure. Proof: Let there be a set A, and let s = sup A, as given. If s is in A, the theorem is proved, and s must be maximum. So consider the case where s is not in A. Then consider an open interval (an open neighborhood) of s : (a,b) which contains s, a < s < b. We first show that (a,s) cannot be empty : If (a,s) is empty then a must be the supremum of A, contradicting our assumption than s is a supremum. Second, since (a,s) is not empty, it must contain a point of A, otherwise s cannot be the supremum again. So it follows that every neighborhood of s contains a point of A, and s is a limit point of A and belongs to $\bar A$.

**Theorem (Product and Box Topologies)**: Given a **cartesian product** $\prod_{\alpha}X_{\alpha}$ of some collection of indexed sets, we can naturally a definition of each point in the product set as a tuple, containing a point from each of the underlying sets. We can define a **projection** function $\pi_{i}(x) = x_i$ - giving the co-ordinate of each point in the underlying "factors". Then the inverse image of $\pi_{i}(U)$ = U x (product of all other sets). 

We can take two types of topologies on the cartesian product :

**Box Topology** : This has as a basis the cartesian product of all open sets of each factor
**Product Topology** : This has as a **sub-basis** the inverse image of an open set in each factor under the projection function to that factor. Thus a finite intersection of any of these will form an open set.

It is immediately obvious that for finite indexed sets, the Box and Product topologies are identical. But in the infinite case, the box topology can have an infinite number of "empty sets" - while the product topology will have only a finite. Thus the box topology is finer than the product topology for infinite products.

**Theorem: Let f: A -> $\prod_{\alpha \in J}X_{\alpha}$, with the product topology, be a function such that $f(a) = (f_{\alpha}(a))_{\alpha \in J}$. Then the function f is continuous iff $f_{\alpha}$ is continuous.**  

**Definition (Hausdorff Space): Any two distinct points have disjoint neighborhoods.**


# Metric Spaces



A **metric** is a function on a set X such that, given and x,y,z, in X :

d(x,y) >= 0 (non-negativity), with equality only if x = y  
d(x,y) = d(y,x) (symmetry)  
d(x,z) <= d(x,y) + d(y,z) (triangle inequality)  

**Open Ball centered at x, or radius r** B(x,r) = {y | d(y,x) < r}, r > 0

**Theorem: If B(x, e) is an open ball, and y is in this ball, then there is a ball B(y,d) which is entirely contained in B(x,e).**  Choose d < e - d(x,y). Then, for any point z in B(y,d), d(x,z) <= d(x,y) + d(y,z) < d(x,y) + e - d(x,y) = e. Thus z is in B(x,e).

**Theorem (Metric Topology)**  If d is a metric on a set X, then the collection of all open balls in X is a basis for a topology in X. 

Proof: Clearly, union of all the balls is in X, and any union of any number of balls is also in X. Consider an intersection of any two balls - B(x1,e1) and B(x2,e2). Any point y in the intersection has a ball centered on y, contained entirely in each of these balls => the smaller ball is in the intersection, and contains y. Thus the collection of open balls induces a topology on X.

If X is a topological space, X is said to be **metrizable** if there exists a metric on set X that induces the given topology. A **metric space** is a metrizable space together with a specific metric d that gives the topology of X.


**Theorem (Hausdorff Spaces - Derived)** Every simply ordered space is Hausdorff in the order topology. The product of  Hausdorff spaces is a Hausdorff space in either box or product topologies. A subspace of a Hausdorff space is a Hausdorff space.


In general, topology studies properties of a space that are independent of a particular metric. This means that some properties, like boundedness, are not considered "topological" properties.

If X is a metric space, a subset A of X is said to be **bounded** if given any two points a and b of A, d(a,b) <= M, where M is a real number. The **diameter** of A is $sup\; d(a1,a2)$ - where a1, a2 are in A. Boundedness is not a topological property. For example, consider the **standard bounded metric** d'(x,y) = min(d(x,y), 1). This induces the same topology on a metric space as the original metric. 

**Theorem: Given two topologies T and T' induced by metrics d and d' : T' is finer that T iff for each x in X, given e > 0, there is a t, such that $B_d'(x,t) \in B_d(x,e)$**  This follows immediately from definitions of "finer" and the two topologies.

**Theorem: (Subspaces of metric spaces) If A is a subspace of a metric space X with metric d, the restriction of d to A x A is a metric for the subspace topology of A.**  

Proof: Let d' = restriction of d to A. Then we see that $B_{d'}(x,r) = B_d(x,r) \cap A$ i.e. the intersection of any basis open set in X with A is a basis open set in A. 

**Theorem: (Metric Spaces are Hausdorff)** Proof : Given x, y, B(x,d(x,y)/2), and B(y,d(x,y)/2) are disjoint.

## Continuity and Metrizability

**Theorem: (Epsilon-Delta Definition) : Let f: X -> Y, and X and Y be metrizable with metrics $d_X$ and $d_Y$. Then f is continuous iff given $x \in X$ and given $\epsilon > 0$, there exists $\delta > 0$ such that :**

$d_X(x,y) < \delta => d_Y(f(x), f(y)) < \epsilon$  

Proof:  
(=>) Let f be continuous. Given $U = B_{d_Y}(f(x), e)$, $f^{-1}(U)$ is open => there is a $\delta$ such that $B_{d_X}(x,\delta)$ is in U.  

(<=) Let the epsilon-delta condition be true. Let U be any set in f. $V = f^{-1}(U)$ should be open. Let for any x belonging X, there is a ball $B_{d_Y}(f(x), e)$ contained in U. And there must be a ball $B_d_X(x,\delta)$ whose image is within the e-ball around f(x), by the epsilon-delta condition. Thus $B_{d_Y}(f(x), e)$ must be in V. Since this is true for every x for which f(x) is in U, V is a union of open balls, and hence open. Thus f is continuous.

**Theorem: (The Sequence Lemma): Let X be a topological space, and A be a subset of X. If there is a sequence in A converging to x, then $x \in \bar A$. If X is, in addition, a metric space, then if $x \in \bar A$ => there is a sequence in A converging to x.**

(=>) Follows immediately from definition of sequence and limit point
(<=) Assume $x \in \bar A$, and X is a metric space. Create a sequence of balls of radius 1/n, in each case pick one point A within each ball (repeats are allowed). Clearly, given any neighborhood of x contains all but a finite no of elements of the sequence, so x is a limit.

Note: We relied on the fact that we could pick nested balls because of the metric, to make this work.

Note : A space is said to have a countable basis at point x, if there is a countable collection ${U_n}_{n \in Z}$ of neighborhoods of x, such that any neigborhood U of x contains at least one of the sets ${U_n}$. A space is said to follow the first countability axiom if every point has a countable basis. Then instead of a ball of radius 1/n, we use $B_n = U_1 \cup U_2 \cup ... \cup U_n$.

**Theorem: (Sequence in functions): Let f: X -> Y be a continuous function, then for every convergent sequence $x_n \to x$, $f(x_n) \to f(x)$. If X is a metrizable, if $x_n \to x => f(x_n) \to f(x)$, then f is continuous.** Note Y need not be metrizable.

(=>) If U is an open neighborhood of $f(x_n)$, and $V = f^{-1}U$, then V is open. V contains x, and hence contains all but a finite number of points of the sequence (for all n > N, $x_n$ is in V). It follows that $f(x_n)$ is in U, and f(x) is the limit of $f(x_n)$.

(<=) Assume $x_n \to x => f(x_n) \to f(x)$. We show that $f(\bar A) \subset \overline{f(A)}$. Let A be any set in X. Let x be in $\bar A$. Since X is a metric space, there is a sequence $x_n$ in A converging to x. Hence $f(x_n)$ converges to f(x), by hypothesis => $f(\bar A) \subset \overline{f(A)}$. Hence f is continuous.


**Theorem: The addition, subtraction, and multiplication functions are continuous functions from R x R -> R. The quotient function is a continuous function from R x R - {0} -> R.**  

\+ : Let p be any point of R, and x be any other point. Then if ${x_n}$ tends to x, we need to show $x_n + p -> x + p$. Given an e > 0, if $|x_n + p - x + p| < e => |x_n - x| < e$.

Similar proofs suffice for -, * and /.


**Theorem: If f, g are continuous functions from X -> R, then f + g, f - g, f * g are continuous. If g(x) != 0 for all x, then f/g is continuous**  Proof. Follows from the previous theorem, and the fact that if h and j are continuous h o j is continuous.


**Definition : Uniform Convergence**  Let $f_n: X -> Y$ be a sequence of functions from set X to a metric space Y. We say the sequence ${f_n}$ converges uniformly to a function f, if given e > 0, there is an integer N such that $d(f_n(x), f(x)) < e$ for all n > N, for all x in X. Note that uniform convergence depends on the metric on Y.

**Theorem: Uniform Limit Theorem: Let $f_n: X -> Y$ be a sequence of continuous functions from set X to a metric space Y. If $\{f_n\}$ converges uniformly to f, then f is continuous.**  
Proof: Use triangle inequality and the epsilon-delta argument.




# Connectedness

**Definition (Separation and Connected Space)** A separation of a topological space X is a pair of disjoint, non-empty, open subsets of X whose union ix X. A space is said to be connected if it has no separation.

Connectedness is a topological property as it is formulated entirely in terms of open sets i.e. any homeomorphism preserved connectedness.

**Theorem: A space X is connected iff the only open subsets of X that are both open and closed are X and the empty set.**  Assume some non-empty proper subset B of X is open and closed. Then $B^c$ is also clopen => X is separated.

**Theorem: If Y is a subspace of X, there is a separation of Y iff there is a pair of disjoint non-empty sets A and B whose union is Y, neither or which contains a limit point of the other.**

Suppose Y has a separation composed of A and B - then A and B are open and closed in Y. We have $\bar A \cap B$ and $A \cap \bar B$ are empty (closures are taken in X) => B does not contain any limit points of A and vice versa.

Assume Y is a union of disjoint non-empty sets A and B, neither of which contains any limit point of the other => both are closed since they contain their own limit points => both are open as well since complements of closed sets are open. So A and B form a separation, being disjoin, non-empty, open sets whose union is Y.

**Theorem: If the sets C and D form a separation of X, and if Y is a connected subspace of X, then Y lies entirely within either C or D.**  

Proof: Since C and D are open, $c \cap Y$ and $D \cap Y$ are disjoin,open, and union is Y => they form a separation of Y if both are non-empty. So one must be empty.

**Theorem: The union of a set of connected subspaces with a point in common is connected.**  
Proof: If they were not, there would be a separation, and the common point would be in one of the disjoint sets forming the separation. And this would force every connected subspace to be in the same set (by the previous theorem), leaving the other empty.

**Theorem: If A is a connected subspace of X, and $A \subset B \subset \bar A$, then B is connected.**  

Proof: Supposed C and D were to form a separation of B. Then A would be all in C, say (wlog). Then $\bar A \subset \bar C$,and $B \subset \bar A$ (by hypothesis). So D is empty - a contradiction.

**Theorem: The image of a connected space under a continuous function (map) is connected.**  
Proof: Let f:X -> Y, with the image f(X) = Z. Then we can consider the surjective function g: X -> Z - g is continuous as it is a restriction of f. Assume Z is separated into A and B. Then their inverses would be open, non-empty (g is surjective) and disjoint (B = $A^c$. $f^{-1}(A^c) = (f^{-1}(A))^c$). Their union is X by definition since every x in X falls in one of the two images. Thus the inverse images form a separation, which is a contradiction as X is connected.

**Theorem: A finite cartesian product of connected spaces is connected.**  
Proof: 
(1) For 2 sets X and Y : x x Y and y x X are connected, being homeomorphic to Y and X respectively. Also consider a point (a,b) in X x Y. $T_x = (X x b) \cup (x x Y)$ is connected, since point (x,b) is common and each is connected. Each $T_x$ contains (a,b). Thus, X x Y = $\cup T_x$ is connected, being the union of connected spaces which have a point in common.
(2) $X_1$ x $X_2$ x ... x $X_n$ is homeomorphic to $X_1$ x $X_2$ x ... x $X_{n-1}$ x $X_n$. So by the previous result and induction, this is proven.

**Definition (Linear Continuum)** : A simply ordered set L having more than one element is called a linear continuum if the following hold :

(1) L has the least upper bound property  
(2) If x < y, there is a z such that x < z < y (dense order)

Obviously, R is a linear continuum.

**Definition (Convex Set in L)** : A subset Y of a linear continuum L is convex if for each pair of points a < b in L, the entire interval (a,b) of points of L lie in Y. Thus all intervals and rays are convex in L.

**Theorem : If L is a linear continuum in the order topology, then L is connected, and so are intervals and rays in L. No other subset of L is connected.**  
Proof: We show that any convex set in L is connected. Consider a convex set Y of L. Assume Y has a separation A and B. Choose a in A, and b in B. Then the set \[a,b\] lies in L. The sets $A_0 = A \cap [a,b]$ and $B_0 = B \cap [a,b]$ form a separation of \[a,b\] because they are disjoint, non-empty, open in \[a,b\] and their union is \[a,b\]. Since A has an upper bound (b), it has a least upper bound (c), since L is a linear continuum. 

If c is in $B_0$, then a < c <= b. Since $B_0$ is open, we have an interval (d,c], d < c, contained in $B_0$ since $B_0$ is open, and (c,b] does not intersect $A_0$, because c is sup A. Then (d,b] is not in $A_0$, and d < c => d is an upper bound of $A_0$, smaller than c, a contradiction.

If c is in $A_0$, we can go the other way : $A_0$ is open, $[c,d)$ is in $A_0$ => d > c is in $A_0$ and $A_0$ is open. There exists a z, such that c < z < d is in $A_0$, because L has dense order. z is in $A_0$, and greater than c, contradicting the assumption c is sup $A_0$.

**Theorem (Intermediate Value Theorem)**: Let $f:X \to Y$ be a continuous function where X is a connected space and Y is an ordered set in the order topology. If a and b are two points of X and if r is a point of Y lying between f(a) and f(b) then there exists a point c of X such that f(c) = r.

Note f(X) is connected, since X is connected and f is continuous. Assume r exists as specified. Let $A = f(X) \cap (-\infty,r)$ and $B = f(X) \cap (\infty, r)$  are disjoint, non-empty (as one contains f(a) and the other f(b)). Each is open in f(X), being an intersection of the open ray with f(X). If r is not in f(X), this would mean A and B are a separation of f(X). So there must be a c in X such that f(c) = r.

**Definition: (Components)** : Define an equivalence relation in a topological space X : Two points are related if they belong to the same connected subspace of X. These equivalence classes are called components of X.

**Theorem (Components are a Disjoint Union of Connected Subspaces)** : Components are connected subspaces, and form a disjoint union of the topological space.

Given two points x and y in space X, a **path** from x to y is a map of a continuous function f from a segment \[a,b\] of R to X, such that f(a) = x and f(b) = y. A space X is **path connected** if every pair of points in X are path connected. **path components** are equivalence classes formed by mutually path connected points.

**Theorem: (Path Connected => Connected): If X is path connected, it is also connected.**

Suppose A and B were a separation of X. Choose x in A, and any two other points a and b. Let A and B be two paths from x to a, and x to b. These paths are connected, so both lie in A. Since a and b are any two points, this means all of X is in A and B is empty - a contradiction.

**Theorem (Path Components are a Disjoint Union of Path Connected Subspaces)**

## Compactness


A topological space is said to be **compact** if every open cover has a finite subcover.

Assume X is a topological space, and there is a subset C of X which has a property that every open cover has a finite subcover. Then we can see, by definition, that C is a **compact space** under the induced topology, since in the induced topology the intersection of each open set in X with C is open in C. So any open cover of C with the induced topology also has a finite subcover.  

Compactness is a strange concept - some examples :

R - the real line - is not compact (ok)  
(0,1) - not compact?
\[0,1\] - compact  (ok)  
R U $\{-\infty, \infty\}$ is compact!  

So compactness is not about size as we think of it normally. Or denseness in a colloquial sense - (0,1) and \[0,1\] are both seem equally dense. Instead we say that in terms of units we care about (i.e. open sets), we can take a collection of them, and spread them to cover the whole space. Now, if we have just one open set containing the "ends", then we can thread our way through the rest till we get to the other end, and this will always be a finite number, if the set is compact. In (0,1), because of the "fuzzy" boundary we can create a series of infinitely reducing open sets such that we never get a finite covering. E.g. Take the sets (1/4n, 1/n), n = (1,2,...). These sets get smaller and smaller, so we can never get a finite subcover. But consider $\[0,1\]$ - you just add one set containing 0 and one containing 1, of some radius e, then we find the same sets will have a finite subcover e.g the set where 1/4n < e

**Theorem: Compactness is a topological property**  
Proof: If h: X -> Y is a homeomorphism, and X is compact, then so is Y. Since $h^{-1}$ is continuous, if O is open in X, h(O) is open in Y. Given an open cover of X, it has a finite subcover, and obviously (since h is onto), also is a finite subcover of Y.

**Theorem: If Y is a compact subspace of X iff any covering of Y by open sets of X contains a finite subcover.** This follows immediately, because any such cover is also a cover of open sets of Y.

**Theorem: Every closed subset of a compact space is compact.**  
Let V be a closed subset of a compact space C. Then the complement of V is open. Take any open cover of V by sets of Y and adjoin the complement of V in C. Then this has a finite subcover - if it contains the complement discard it. Then the remaining sets form a finite subcover of the original cover.

**Theorem: Suppose $K \subset Y \subset X$. Then K is compact relative to X iff K is compact relative to Y.**  
(=>) K is compact relative to X - any open cover of K in Y can be extended to an open cover of K in X and hence has a finite subcover.
(<=) K is compact relative to Y - An open cover of K in X can be reduced to an corresponding open cover in Y.

**Theorem: If X is a Hausdorff space, Y is a compact subspace of X and x is a point in X - Y, then there are disjoint open sets covering Y and x respectively.**  
Between x and every point of Y
there are disjoint neighborhoods (Hausdorff). The ones around Y form an open cover and since Y is compact, have a finite subcover. Their union is an open set covering Y. The (finite) intersection of the corresponding neighborhoods around x contain a disjoint open set containing x.

**Theorem: Every compact subspace in a Hausdorff space is closed.**  Proof: If X is the space, Y is the compact subspace, X - Y is open, because every x in X - Y has an open neighborhood disjoint from Y => X - Y is open and Y is closed.

**Theorem: The image of a compact subpace under a continuous function (map) is compact.**  Take an open covering of the image. The inverse image forms an open cover of the compact space and hence has a finite subcover. The corresponding images form a finite subcover of the image.

**Theorem: Let f be a one-one, onto (bijective) continuous function from a compact space X to a Hausdorff space Y. Then f is a homeomorphism i.e. its inverse is continuous.**  If A is closed in X, then it is compact, and since the continuous image of a compact image is compact, f(A) is compact in Y. Since Y is Hausdorff, f(A) is closed in Y. Hence $f^{-1}$ is continuous, and f is a homeomorphism.

**Theorem: The product of finitely many compact sets is compact**    
Proof: Somewhat tedious proof - Given an open covering of X x Y, prove that for each a x Y, there is a finite subcover of boxes (U x V - U is a basis set of X, V is a basis set of Y). We need a fixed "width" of U's, so we take the intersection of U's i.e. the smallest U across all of Y, which contains a. The a x Y. Now go across X, and prepare more of these "tubes" - the collection forms a cover of X, and hence has a finite subcover. Given this, we have a finite number subcover of X x Y.

**Theorem: Finite Intersection Property (FIP)** A collection C of subsets of X is said to have the finite intersection property if every finite subcollection has a non-empty intersection. This is almost always used for nested intervals.

**Let X be a topologogical space. Then X is compact iff for every collection C of closed sets in X having the finite intersection property, the intersection of all sets in C is non-empty.**  

Proof: Basically, $\cap C_i = X - \cup C_i^c$. Each of $C_i^c$ is an open set. Thus $\cap C_i$ being empty, means the complements form an open cover of X, and by compactness of X, there is a finite subcover. The finite subcover also covers all of X, so this implies the corresponding finite set of closed sets has an empty intersection, violating FIPS. Thus, $\cap C_i$ following FIP cannot be empty.

**Theorem: A nested sequence of non-empty closed sets in a compact space has a non-empty intersection.** Such a sequence has the FIP property. Then we must have a non-empty intersection.

**Theorem (Heine-Borel): Let X be a simply ordered set with the least upper bound property. In the order topology, each closed (and bounded) interval of X is compact.**   
Proof: Basic idea of the proof is that given an interval \[a,b\], we can creep along the interval with a finite number of steps. Assume there is an open covering of the given interval. 
First, given X is simply ordered, for every element x of the interval < b, we can find a y > x of the interval such that the interval \[x,y\] can be covered by at most 2 elements of the open covering. First we note that the set of successors of x is not empty - b is there. If x has an immediate successor, obvious we can cover it with two sets. If x has no immediate successor, then an open set covers x, so we can find a y > x in that set itself.

Now, let S = {x : x in \[a,b\], \[a,x\] is covered by a finite number of sets}. First a is in S, so S is not empty. S is bounded above by b, so S has a supremum, c <= b (because b is an upper bound). Is c in S? If c is not in S, then there is an open set covering c, disjoint from S (otherwise c would be in S, having just one more set than that any point of S in that set). But if the open set is disjoint, there must be a d < c, disjoint from S and c cannot be the supremum. So c must be in S.

Finally, c = b : Because if c < b, then we can find a y, such that c and y are covered by at most 2 sets. So y would be in S, contradicting that c is the sup S. Hence c = b.

**Corollary : Every closed (and bounded) interval of R is compact.**  

**Theorem (Extreme Value Theorem)** Let f: X -> Y be continuous, where Y is an ordered set in the order topology. If X is compact, then there exist points c and d in X such that f(c) <= f(x) <= f(d) for every x in X.

**Definition (Distance from point x to subset A)** : In a metric space, d(x,A) = inf d(x,a) where a is in A.

**Theorem (Distance from set is a continuous function): In a metric space X, given a fixed set A, d(x,A) is a continuous function.**  

Given two points x and y, d(x,A) <= d(x,a) <= d(x,y) + d(y,a) for all a in X. 
Thus d(x,A) <= d(x,y) + d(y,A) => d(x,A) - d(y,A) <= d(x,y) - the equality works both ways :

d(y,A) - d(x,A) <= d(x,y) => |d(x,A) - d(y,A)| <= d(x,y). Apply the epsilon-delta criteria, we get continuity, given an e, such that |d(x,A) - d(y,A)| < e, if we y such that d(x,y) < e => |d(x,A) - d(y,A)| < e. 

**Theorem (Lebesgue Number Lemma) : Let (X,d) be a compact metric space, with an open covering A. There exists a $\delta > 0$, such that each subset of X having diameter less than $\delta$ is contained in some subset of A.**  

Proof: Take an open covering on X, of which we have a finite subcover $A_1, A_2, ..., A_n$. Let $C_i = X - A_i$ be corresponding closed sets. If one of the $A_i$ is X, we are done. For any x in X, $d(x,C_i) > 0$, for some $C_i$, as each $C_i$ is closed, and x is not contained in at least one. 

Thus, if we take the function $f(x) = \frac{1}{n}\sum_{i=1}{n}d(x,C_i)$, f(x) > 0 for all x.

Also f(x) is continuous, because we have proven that the sum of and division by non-zero continuous functions is continuous, and the distance from a set is continuous.

Thus f(x) attains a minimum value - $\delta$(as X is compact). The minimum cannot be zero, as we have seen, since f(x) > 0 for all x. 

Now, given any subset B of X of diameter less than $\delta$. Take a point $x_0$ of B. Then B lies in a $\delta$ neighbourhood of $x_0$. Thus $\delta <= f(x_0) <= sup d(x_0, C_i)$. Thus the $\delta$ neighborhood of x is contained in some $A_m$, corresponding to the $C_m$ such that $d(x_0, C_m)$ attains the maximum.

Why study separation of points? In most analysis / topology problems, we often have two notions :

(1) Any two points are separated / distinct or far apart in some sense  
(2) Every point can be approached very closely from other points, and a sequence of points will usually cluster around a single point  

Basically we need separation, but also connectedness. And "good" topological spaces have both qualities. 

Given any two distinct points a and b, you may have the following situations :

(1) Every open set containing a contains b, but b has open sets containing it, that do not contain a - so a sticks to b in some sense. a sticks to b, b doesn't always carry along a.

(2) Every open set containing a contains b, and every open set containing b contains a - then the two points are **topologically indistinguishable**, and the relation is symmetric, transitive and reflexive i.e. an equivalence relation. 

(3) Two points may have open sets around them that not only contain the other point, but are also disjoint. This means we can separate not only the points, but even the space surrounding them in some sense.



A **sequence of points** in a space X is a function f: N -> X. The **limit of a sequence or a sequential limit** is a point in x such that any neigborhood of the point contains all but finitely many points of the sequence. Note that sequences (and functions) have limits. A sequence of points approaches a limit point arbitrarily close, infinitely often. For the limit of a sequence, on the other hand, the sequence approaches and then gets trapped within a smaller and smaller neighborhood and never leaves that neighborhood. 

A sequence may have repeated values - if a sequence ends with infinitely many constant values, then the sequence may not be a limit point of a set containing it. But the limit of a sequence is always the limit point of a set containing the sequence if *it is not itself in the sequence*.

**$T_0$ Space (Kolmogorov Space)** No two points are indistinguishable. Then if we define a < b, if every open set of a contains b, then this relation is a partial order. Any space can be converted to a $T^0$ space, by defining a topology over the equivalence classes of points which are topologically indistinguisable.

**$T_1$ Space** Given any two points a and b, there is an open set containing a but not b, and another containing b but not a. 

Another way to say this is that each set {a} is closed. Proof: Given a in X, taken the union of open sets around them not containing a. This is open and equal to X - {a}, and open, so {a} is closed.

**$T_2$ Hausdorff Space** Any two distinct points have open sets containing them that are disjoint.

Hausdorff spaces have a most important property :

**Theorem: In a Hausdorff space, every sequence has at most one limit.** 

```
Note the reverse is not necessarily true i.e. a space may be such that every sequence has at most one limit, but not be Hausdorff. It does have to be T1 though, as proved later.
```

Proof(1): Assume there are two limit points a and b. Since a is a limit point, there is an open set around a, outside of which there are only finitely many points of the sequence. There is an open set around b, disjoint from a, where there are again only finitely many points of the sequence. b has an at least one open set which does not contain each of these points (because the space is Hausdorff), and their intersection is open since they are finite (because of topological axioms), so there is an open set around b which does not contain any point of the sequence, contradicting the assumption that b is a limit point of the sequence.

Proof(2): Assume there are two limit points a and b, with two disjoint open sets U and V around them. Then there are only finitely many points of the sequence outside of each - and this number cannot be zero since each of U and V contains at least one point, being limits of the sequence. Thus $U^c$ and $V^c$ contain finitely many points of the sequence i.e. $U^c \cup V^c$ also contains only finitely many points. This implies that $(U^c \cup V^c)^c = U \cap V$ contains all but a finite number of points of the sequence i.e. it is not empty, contradicting our original assumption.

**Theorem: In a space, if every sequence has at most one limit, then it is a T1 space**  
Proof: Given any two points a and b, consider the sequence {b, a, a, ...}. If any open set containing a contained b, then both a and b would be limits. Since this is not possible, the limit can only be a i.e. there is an open set containing a but not b. Similarly, we can prove the other way.

Taken n neighborhoods of x, each not containing one of the points - their intersection is open (since finite intersection of open sets is open), so we have an open neighborhood of x containing no point of A.

**Convergent Sequences** : A sequence ${p_n}$ in a metric space X is said to converge if there is a p in X with the following property : For *every* $\epsilon$ > 0, there is an integer N, such that $d(p,p_n) < \epsilon$ for all n >= N.

We say, ${p_n}$ **converges** to p, or p is the **limit** of ${p_n}$. We write $p_n \to p$ or $lim_{n \to \infty} p_n = p$.

If ${p_n}$ does not converge, it is said to **diverge**.

Note that a sequence may converge is one space and not another - ${1/n}$ converges in R, but not in the set of all positive real numbers.

## Separation of Sets and Points

References :

[Topology - The Journey Into Separation Axioms](https://www.cmi.ac.in/~vipul/mathjourneys/contytopologysep.pdf)

[Wikipedia on Separated Sets](https://en.wikipedia.org/wiki/Separated_sets)

Why study separation of points? In most analysis / topology problems, we often have two notions :

(1) Any two points are separated / distinct or far apart in some sense  
(2) Every point can be approached very closely from other points, and a sequence of points will usually cluster around a single point  

Basically we need separation, but also connectedness. And "good" topological spaces have both qualities. 

Given any two distinct points a and b, you may have the following situations :

(1) Every open set containing a contains b, but b has open sets containing it, that do not contain a - so a sticks to b in some sense. a sticks to b, b doesn't always carry along a.

(2) Every open set containing a contains b, and every open set containing b contains a - then the two points are **topologically indistinguishable**, and the relation is symmetric, transitive and reflexive i.e. an equivalence relation. 

(3) Two points may have open sets around them that not only contain the other point, but are also disjoint. This means we can separate not only the points, but even the space surrounding them in some sense.

A **sequence of points** in a space X is a function f: N -> X. The **limit of a sequence or a sequential limit** is a point in x such that any neigborhood of the point contains all but finitely many points of the sequence. Note that sequences (and functions) have limits. A sequence of points approaches a limit point arbitrarily close, infinitely often. For the limit of a sequence, on the other hand, the sequence approaches and then gets trapped within a smaller and smaller neighborhood and never leaves that neighborhood. 

A sequence may have repeated values - if a sequence ends with infinitely many constant values, then the sequence may not be a limit point of a set containing it. But the limit of a sequence is always the limit point of a set containing the sequence if *it is not itself in the sequence*.

**$T_0$ Space (Kolmogorov Space)** No two points are indistinguishable. Then if we define a < b, if every open set of a contains b, then this relation is a partial order. Any space can be converted to a $T^0$ space, by defining a topology over the equivalence classes of points which are topologically indistinguisable.

**$T_1$ Space** Given any two points a and b, there is an open set containing a but not b, and another containing b but not a. 

Another way to say this is that each set {a} is closed. Proof: Given a in X, taken the union of open sets around them not containing a. This is open and equal to X - {a}, and open, so {a} is closed.

**$T_2$ Hausdorff Space** Any two distinct points have open sets containing them that are disjoint. 

Hausdorff spaces have a most important property :

## Irreducibility 

A topological space is said to be "irreducible" or "hyperconnected" if it cannot be expressed as a union of two proper closed subsets.

Theorem: A topological space X is irreducible if there is no pair of disjoint proper open sets.
=> We prove the contrapositive. Let U and V be open, disjoint proper subsets in X (if there are none, we have trivially proved the theorem), then we could take the intersection of all the other sets with V => each would be open. And their union would be open too. And hence U and X-U are both open, they are also both closed which implies that X is, in fact, not irreducible.

If no two proper open sets are disjoint then every open set is dense in the another, because every point of an open set has a open neighborhood entirely in that set, and each such open neighborhood must contain a point of every other open set! That is why the word hyperconnected is used. Every neigborhood of every point contains a point of every other open set.
