Definition : A set X is a metric space if it has a notion of distance, if there exists a metric d from X2  to R such that for any p,q in X :
i) d(p,q) >= 0 (non-negativity), d(p,q) = 0 means p = q.

Neighborhoods : Open and Closed Subsets of a Metric Space

In a metric space, a neighborhood of a given point is a the set of points “near” or within a certain distance of a given point. The set of points within a radius < r (where r is positive) around a point forms a neighborhood also known as an open ball. 

An open ball : neighborhood : Nr = { y : d(x,y) < r }

A closed ball has a radius exactly r, and thus includes the “boundary” points at distance r. 
Nr-bar= { y : d(x,y) <= r }

In other definitions, a neighborhood of a point or a set is often defined as a set that contains an open set containing the point or set. In this case neigbhborhoods could also be closed.

Neigborhoods and Points

A point is said to be a limit point of a set, if any neighborhood of the point contains at least one point of the set, excepting itself.

An interior point of a set is one with at least one neighborhood containing only points of that set (including itself). 

A point belonging to a set and not a limit point of the set is called an isolated point, since we can see that it has a neighborhood where no other point of that set exists.

G = { 1/n : n in N} : 0 is a limit point. All other points are isolated.

Open and Closed Sets

A set in a metric space is an open set if every point of the set is an interior point.

A closed set contains all its limit points.

A closure of a set is the set together with all its limit points - often referred to by a bar over the set. (Closure of E is E-bar).

A set may be neither open nor closed, or both open and closed.

Theorem : E is closed <=> E is equal to its closure

Proof :  E-closure = E U set of limit points not in E = E U empty-set = E.

Theorem : The closure of a set is the smallest closed set that contains it.

Proof : We prove that any closed set U containing a set E, contains all the limit points of E. If p is any limit point of E, then any neighborhood of p contains a point of E, and hence of U. Hence p is a limit point of U and therefor in U. Thus any closed set containing E contains all the limit points of E also. Their intersection is therefore the closure of E, which is the smallest closed set containing E.

Theorem :  E is open <=> Ec is closed.

Proof : Any point not in Ec has a neighborhood not containing any points of Ec, and hence is not a limit point of Ec . So Ec has all its limit points and is closed.

Theorem :

Proof : We use the fact that : (U E)c = intersection Ec , in the following proof.

Arbitrary intersections of closed sets are closed, because their complements are open, and hence the union of the complements is also open. The complement of the union of complements is the intersection, which is therefore closed.

To prove that a finite union of closed sets is closed. Let p be any point not in the union of these sets. For each closed set, the point P has a neighborhood not containing any point of the closed set. The smallest such set is included in all the others (and exists because the no of closed sets is finite) and forms a neighborhood not including any point in the union of the closed set. Thus any point not in the union is not a limit point of the union, and the union of closed sets is closed. 

Union of the complements of a finite set of open sets is closed, since each complement is closed. The complement of the union, i.e. the intersection of the open sets, is therefore open.

Definitions :
An open cover of a set E in (a metric space) X is a (potentially uncountable) collection of open sets who union contains E. The open cover is said to “cover” E.

A subcover of a cover is a sub-collection of the cover that will also cover E.

A set K is compact (in X) if every open cover of K contains a finite subcover.

Theorem : Let X be a metric space, and Y a subset of X. A set E in Y is open in Y <=> E = Y n G for some G open in X.

Proof : 
<= : We take E = Y n G, where G is open in X. Either a point in E is a limit point of G-E, or it is not. For those which are not, they have a neigborhood entirely in E. For those which are limit points (i.e. any neigborhood contains points of E and G - E), Take the intersection with Y - then this will be an open ball in Y, containing only points in E. Thus there truncated neigborhoods are neighborhoods in Y containing only points in E. Thus every point in E has a neighborhood containing only points in E - so E is open.

=> : We take E as open in Y. Every point x in E has a neighborhood of radius r containing only points in E - say NY(x,r). Hence U NY(x,r) = E. Thus if we take the corresponding neighborhoods in X, NY(x,r), then these will consist of points either in E, or in X - Y. The union of these neighborhoods will be an open set in X (since an arbitrary union of open sets is open), which includes E. If we call this set G, then

Y n G =Y n (U NX(x,r)) = U (Y n NX(x,r)) = U NY(x,r) = E.
Note : This theorem is used in the proof below, to show compact sets can be used as metric spaces in their own right.

Theorem : If Y is contained in X, then K compact in Y <=> K being compact in X.
Proof : 
(<=) : Assume K is compact in X. Let { GY-alpha } be an open cover of K in Y. Derive a { GX-alpha } by taking open sets such that GY-alpha = GX-alpha  n Y (since every open set G in Y in included in an open set G’ in X such that G = G’ n Y).   

This forms an open cover of K in X, . Thus it has a finite subcover, since K is compact in X. The corresponding open sets in Y form a finite subcover of { GY-alpha } - hence any open cover of K has a finite subcover.

(=>) : Assume K compact in Y. Let { GX-alpha } be an open cover of K in X. Then consider {GY-alpha } such that GY-alpha   = GX-alpha n Y. Each GY-alpha  is also open, as mentioned in the previous theorem.

Also, K is subset in U {GX-alpha} and in Y, and hence K is included in Y n U {GX-alpha} = U {Y n GX-alpha} = U {GY-alpha} . 

Thus {GY-alpha} is an open cover of K in Y. Since K is compact, it has a finite subcover - the corresponding collection in X is also finite and covers K, and so is a subcover of K in X.  Hence every open cover of K in X has a finite subcover.

From Rudin : This theorem allows us to use compact sets as metric spaces in their own right, without bothering about the embedding space. In particular, while we dont talk of open or closed spaces, we can talk of compact metric spaces.

Theorem : Compact subsets of metric spaces are closed.

Proof : For each point q in compact set K to any point P not in K, consider an open ball of radius r = d(q,P)/2. This forms an open cover of K and has a finite subcover, of size n, say. 

For this finite subcover consider the corresponding finite collection of neighborhoods (open balls) of P with radius d(q,P)/2. Each of these open balls does not intersect with the corresponding open ball centered on q. Also they are totally ordered by inclusion - hence the intersection of all these sets is the smallest of the neighborhoods, and does not intersect any open ball in the finite subcover of K. Thus it does not contain any point of K i.e. any point in Kc has a neighborhood containing only points in Kc - i.e. Kcis open. Hence K is closed.

Theorem: A closed subset of a compact set is compact.

Theorem : Let P be a non-empty perfect set in Rk. Then P is uncountable.

Proof : Since P has limit points, P is infinite. 

If P is countable, we can denote it as {x1, x2, .., xn}.

We construct a sequence {Vn} of neighborhoods as follows :
i) Let V1 be a neighborhood of x1.
ii) If Vk is a neighborhood of xk, such that Vk n P is not empty. Then we can find a neighborhood Vk+1 such that 
a) Vk+1-closure is included in Vk 
b) xk is not in Vk+1-closure
c) Vk+1 contains points of P.

We can always do this. If xk is in Vk then we can do this by picking another point p of P in Vk+1 and picking a neighborhood contained entirely in Vk, but of radius less than d(p, xk). If xk is not in Vk then we just pick a neighborhood entirely in Vk.

Now consider Kn = Vn-bar n P. Clearly, Kn is not empty, since each Vn contains points of P. Also Vn-bar and P are closed sets, so their intersection Kn is closed, and since Vn is bounded so is Kn. Thus Kn is compact. Thus {Kn} forms a nested sequence of compact sets, with the intersection of any finite number being non-empty. Hence the intersection must be non-empty i.e. n {Kn} must contain a point of P. This immediately leads to a contradiction, because the infinite intersection n { Kn } is empty since for each xn in P, Kn+1 does not contain xn, so no xn is a part of n {Kn}. Hence P is uncountable.

To show that the set of algebraic numbers is countable, let Lk denote the set of algebraic

Basic Proof :

1. Consider 

All polynomials with coefficients a0, a1, a2, … an such that

let p(x) = anxn + an-1xn-1 + an-2xn-2 + … +  a0 . Each such polynomial has a root, and we can easily prove that it has no more than n roots by simple induction on n.

We define the height of a polynomial as 

h(p) = |a0| + |a1| + … + |an| + n 

Let S(k)  = { x : x is set of all solutions of polynomial with height less than k}.

Example : S(1) = {}, S(2) = {0}, S(3) = {0,1,-1}

What is the cardinality of S(N)? Clearly each of |a0|, |a1|, … <= N. 

For each n, the total combinations therefore are less than (2N + 1)n, where n <= N (including negative values and zero). Since n can be from 1 to N, we can say the total number of combinations (distinct equations) is less than N.(2N+1)N . Each equation will have no more than n roots, as can be easily proved using induction. Hence the number of roots (cardinality of S(N)) is no more than N2(2N + 1)N which is a finite number.

Since every integer i is a root of a polynomial (x = i), hence the algebraic numbers are infinite in number. 

Also {S(k)} form a countable sequence of finite sets containing the solutions to all polynomial equations. We can form a sequence {S’(k)} where each S’(k) = S(k) - U S(j), where 1 <= j <= k-1, to eliminate duplicates. Since each S(k) is finite, so is each S’(k), and the union of {S’(k)} is the set of all algebraic numbers which are there countable.
