4.1 Definition (Limit of a function) : Given a function from a set E in a metric space X to a metric space Y, we say f(x) -> q for x -> p if given e, there is a delta D such that d(x,p) < D, implies d(f(x), q) < e. 

Note : p need not be in E. Also, p may even be an isolated point of E - in this case the limit always exists and is f(p).

4.2 Theorem (Sequence Characterization) : lim (x->p) f(x) = q ⇔ for all sequences {pn} in E with lim pn = p ( pn <> p ), lim f(pn) = q.

Proof :
( => ) From the hypothesis, given an e > 0, there exists D > 0 st 0 < d(x,p) < D => d(f(x), q) < e 
Given a sequence {pn} converging to p, there is an N such that d(pn, p) < D for all n >= N. Hence d(f(pn), q) < e by the hypothesis i.e. q is the limit of f(pn).

( <= ) Not (A) => Not(B) means B => A. We assume lim (x->p) f(x) <> q. We have to prove there is at least one sequence {pn} converging to p, but lim f(pn) <> q.
By hypothesis, there is an e > 0, s.t. for all D > 0, there is an x in E st, 0 < d(x,p) < D, but d(f(x), q) > e. 
We propose a sequence Dn = 1/n, and choose xn such that d(xn, p) < Dn, but d(f(xn), q) > e. Hence {xn} converges to p, but lim f(xn) <> q.

Corollary : If f has a limit at p, it is unique.
Otherwise we can construct two sequences converging to p, with the function converging to two different limits, contradicting theorem 4.2.

4.4 Theorem : Suppose E is in X (a metric space), p is a limit point of E, f and g are complex functions on E, and 

lim f+g(x) = lim f(x)  + lim g(x)
lim f*g(x) = lim f(x)*lim g(x)
lim f/g(x) = lim f(x) / lim g(x) if lim g(x) <> 0.

Proof : From the sequence characterization of functions, these properties follow immediately.
This is true for C, R
For Rk we have f+g and f*g as vector addition / inner product / scalar multiplication. f/g is not defined.

4.5 Definition : Continuity of functions :

X, Y are metric spaces. p is a point in E in X, and f is a function from E to Y.
We say f is continuous at p if : there is an e > 0, there exists a D such that for all x in E,

d(x, p) < D => d(f(x), f(p)) < e.

Note : p is a point in E. Thus, as far as continuity goes, we are only concerned with points of E, and we can ignore X for this purpose. Does this mean E has to be closed? No. So if x -> q not in E, we cannot say anything about continuity at q. If p is an isolated point, f is always continuous at p.

4.6 Theorem : If p is a limit point of E, f continuous at p ⇔ lim (x -> p) f(x) = f(p)
Note : If E has no limit points, then f is continuous vacuously. Somewhat non-intuitively, any function on a discrete metric are continuous.

4.7 Theorem : Composition of two continuous functions is continuous.
X, Y, Z are metric spaces. E is a subset of X. f is a function from E to Y. g is a function from f(E) to Z. we define h(x) = g(f(x)) as a function from X to Z. Then h(x) is continuous from theorem 4.6. 
We write h = g o f. 

Note : Because of this, and the previous theorem, we can also say : lim x -> p f(g(x)) = f (lim x->p g(x)), if f, g are continuous.

4.18 Definition : Uniform Continuity

A function f from a metric space X to Y is said to be uniformly continuous if : given e > 0, there exists a d > 0 such given p and q in X, dx(p,q) < d => dy(f(p), f(q)) < e.

4.19 Theorem : Let f be a continuous mapping of a compact metric space X into a metric space Y. Then f is uniformly continuous on X.

Proof : Given e > 0, let D be a real-valued function on X such that for any p in X,

dx(p, x) < D(p) for any x in X => dy(f(p), f(x)) < e/2.

Take the collection of sets {Vp} = N(p, D(p)/2). Clearly these form an open cover of X and so have a finite subcover {Vpi} i = 1 to m, say, for the points p1, p2 .. pm.

Let d = ½ min(N(pi)), i = 1 to m.

Consider any p, q  such that d(p, q) < d. Now p belongs to some N(pm, D(p)/2) i.e. 
dx(p, pm) < D(p)/2. Thus, by the triangle inequality :

dx(pm,q) <= dx(p,q) + dx(p, pm) < d + D(p)/2 <= D(p). 

Thus dy(f(pm), f(q)) < e/2, and also, dy(f(pm), f(p)) < e/2. 

We get dy(f(p), f(q)) <= dy(f(p), f(pm)) + dy(f(pm), f(q)) < e.
