```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.init_printing()

a = sp.symbols('a', real=True, nonzero=True)
th = sp.symbols('theta', real=True)
f = sp.Function('f')
g = sp.Function('g')
u = sp.symbols('u')

x = sp.symbols('x')


def plot2d(ax: plt.Axes, param, range, xf, yf, *, yaspect: float = 1.0, debug=False, color='blue', fmt='-', label=None):
    param_range = np.linspace(range[0], range[1], range[2])
    f = sp.lambdify((param), xf)
    if debug:
        display(xs)
        display(ys)
    xs = f(param_range)
    g = sp.lambdify((param), yf)
    ys = g(param_range)
    ax.tick_params(axis='both', which='major', labelsize=8)
    ax.set_xlabel(xf)
    ax.set_ylabel(yf)
    ax.set_aspect(yaspect, anchor='W')
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')
    ax.grid()

    ax.plot(xs, ys, fmt, color=color, label=label)
```

```python
# x sin(1/x)

fig, axes = plt.subplots(1,1 ,figsize=(16,6))

plot2d(axes, x, (-np.pi/20, np.pi/20, 200), x, x, color='pink', fmt="--")
plot2d(axes, x, (-np.pi/20, np.pi/20, 200), x, -x, color='pink', fmt="--")
plot2d(axes, x, (-np.pi/20, np.pi/20, 200), x, x*sp.sin(1/x))
```

```python
fig, axes = plt.subplots(1,1 ,figsize=(10,16))

plot2d(axes, x, (-1, 1, 100), x, x**2, color='pink', fmt="--")
plot2d(axes, x, (-1, 1, 100), x, -(x**2), color='pink', fmt="--")
plot2d(axes, x, (-np.pi/4, np.pi/4, 1000), x, x*x*sp.sin(1/x))

axes.set_ylim(-.5,.5)
```

```python
fig, axes = plt.subplots(1,1 ,figsize=(8,10))

plot2d(axes, x, (-1, 1, 100), x, sp.sqrt(sp.Abs(x)), color='pink', fmt="--")
plot2d(axes, x, (-1, 1, 100), x, -sp.sqrt(sp.Abs(x)), color='pink', fmt="--")

plot2d(axes, x, (-1, 1, 1000), x, sp.sqrt(sp.Abs(x))*sp.sin(1/x))
```
