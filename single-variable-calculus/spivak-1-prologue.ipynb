# Basic Properties of Numbers

Numbers have two operations defined: sum (+) and multiplication (+). 
For any two numbers, their sum a + b, is also a number, and a.b is also a number.

## Rules for addition

**(P1) Associative law for addition : a + (b + c) = (a + b) + c**

This allows us to unambiguously write a + b + c + d 

**(P2) Existence of an additive identity: If a is any number, a + 0 = a**

**(P3) Existence of additive inverses: For any number a, there is a number -a, such that a + (-a) = (-a) + a = 0**

Using (P3) we can prove **0 is unique**. 

Given a + x = a

Hence, (-a) + (a + x) = (-a) + a = 0 (definition of sum, P3)

Hence ((-a) + a) + x = 0 (P1)

Hence, 0 + x = 0 (P3)

Hence, x = 0 (P2)

We also define **subtraction** as derived from addition:

a - b = a + (-b)

We can also easily show that **cancellation** works for addition:

a + b = c + b => a = c

**(P4) Commutative law for addition: a + b = b + a**

The order of addition does not matter

## Rules for multiplication

**(P5) Associative law for multiplication: a.(b.c) = (a.b).c**

**(P6) Existence of a multiplicative identity: a.1 = 1.a. Moreover $1 \ne 0$**

**(P7) For every number $a \ne 0$, there is a number $a^{-1}$ such that $a.a^{-1} = a^{-1}.a = 1$**

Note that we can define **division** in terms of multiplication:

$a/b = a.b^{-1}$ provided $b \ne 0$

**cancellation** works for multiplication, but not always:

if a.c = b.c, then a = b provided $c \ne 0$


**(P8) If a and b are any numbers, a.b = b.a**

**(P9) Distributive Law: a.(b + c) = a.b + a.c**

**When a - b = b - a**

Given: a - b = b - a => (a - b) + b = (b - a) + b = b + (b - a)
a = b + (b - a)

a + a = (b + b - a) + a = b + b
a.(1 + 1) = b.(1 + 1)
a = b

**Multiplication by zero: $a.0 = 0$**

a.0 + a.0 = a.(0 + 0) = a.0

Hence, a.0 = 0

**Prove (-a).b = -(a.b)**

(-a).b + a.b = (-a + a).b = 0.b = 0

**Prove (-a).(-b) = ab**

(-a).(-b) + (-a.b) = -a.(-b + b) = -a.0 = 0

Hence, (-a).(-b) = ab

## Ordering & Closure Properties

**(P10) Trichotomy law**

For every number a, one and only one of the following holds:

i) a = 0

ii) a > 0 (a is positive)

iii) -a > 0 (a is negative)

**(P11) Closure under addition for positive numbers: if a > 0 and b > 0, then a + b > 0**

**(P12) Closure under multiplication for positive numbers: if a > 0 and b > 0, then a.b > 0**


**Definitions**

a > b if a - b > 0

a < b if b > a

a >= b if a > b or a = b

a <= b if a < b or a = b


Proofs: 

if a and b are any two numbers, one one of the 3 holds:

i) a - b = 0

ii) a - b > 0

iii) -(a - b) = b - a
