# Functional Plot

```python
# linspace - generate 10 evenly spaced numbers from 0 to 5, inclusive
x_1 = np.linspace(0,5,10)
y_1 = x_1**2
plt.plot(x_1, y_1)
plt.title("Days Worked");
plt.xlabel("Days")
plt.ylabel("Days Squared")
```

```python
# multiple plots

plt.subplot(1,3,1)
plt.plot(x_1, y_1, 'r')
plt.subplot(1,3,3)
plt.plot(x_1, y_1, 'b')
```
