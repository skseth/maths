# Subspaces in $R^n$

Generalizing the concept of lines, planes in $R^n$:

A set $V \in R^n$ is called a **subspace** of $R^n$ if it satisfies the following conditions:

a) $\vec{0} \in V$

b) $\vec{x} \in V \implies c\vec{x} \in V$

c) $\vec{x}, \vec{y} \in V \implies \vec{x} + \vec{y} \in V$


**Definition: Linear combination / Span** 

Given $\vec{v_1}, \vec{v_2}, \dots, \vec{v_n} \in R^n$, if $c_1, c_2, \dots, c_n \in R$, then the following expression :

$c_1\vec{v_1} + c_2\vec{v_2} + \dots + c_n\vec{v_n}$

is called a **linear combination** of the vectors $\vec{v_1}, \vec{v_2}, \dots, \vec{v_n}$

The set of all linear combinations of vectors $\vec{v_1}, \vec{v_2}, \dots, \vec{v_n}$ is called their span, Span($\vec{v_1}, \vec{v_2}, \dots, \vec{v_n}$). We say vectors $\vec{v_1}, \vec{v_2}, \dots, \vec{v_n}$ spans a set V, if Span($\vec{v_1}, \vec{v_2}, \dots, \vec{v_n}$) = V.

**Proposition: Let $\vec{v_1}, \vec{v_2}, \dots, \vec{v_n} \in R^n$. Then Span($\vec{v_1}, \vec{v_2}, \dots, \vec{v_n}$) is a subspace of $R^n$**

Proof: 

Let the span be called V.

- 0 is in V, choosing $c_1, c_2, ... , c_n = 0$ we get $c_1v_1 + c_2v_2 + ... + c_nv_n = 0$
- if $x = c_1v_1 + c_2v_2 + ... + c_nv_n$ is in v, then $cx = (cc_1)v_1 + (cc_2)v_2 + ... + (cc_n)v_n$ is also a linear combination and hence in V
- if $x = c_1v_1 + c_2v_2 + ... + c_nv_n$ and $y = d_1v_1 + d_2v_2 + ... + d_nv_n$ are in v, then $x+y$ is also in V

**Definition: Orthogonal Subspaces and Orthogonal complement**

Let $V, W \in R^n$ be subspaces. We say V and W are orthogonal subspaces if for every element $\vec{v} \in V$ and $\vec{w} \in W$, $v.w = 0$

Given a subspace $V \in R^n$, define the it's orthogonal complement: $V^{\perp} = \{  \vec{x} \in R^n: \vec{x} \cdot \vec{v} = 0 \text{ for every } \vec{v} \in V\}$

**Proposition: Given a subspace $V \in R^n$, it's orthogonal complement: $V^{\perp}$ is also a subspace**

Proof: Let v, w in V and x, y in $V^{\perp}$.

- 0 is in $V^{\perp}$, since v.0 = 0 for any v in V
- v.(cx) = cv.x = c0 = 0
- v.(x + y) = v.x + v.y = 0 + 0 = 0


**Idea: Affine Space**

Given a vector subspace $V \in R^n$ and a vector w, the set $A = \{ \vec{x} \in R^n: x = v + w \text{ for some } v \in V\}$ is called an affine space. Note in particular that $\vec{0}$ may not belong to an affine space.

# Ex 5: $Span(v_1, v_2, \dots, v_n)$ is the smallest subspace contain $v_1, v_2, \dots, v_n$

Proof: Let W containing vectors $v_i$ be a subspace. 

Then every linear combination of $v_i$ is in W (this can be proved using induction and the scaling and addition conditions). e.g. if $v_i$ is in W then so is $c_iv_i$. And if $c_1v_1$ and $c_2v_2$ are in W, so is $c_1v_1 + c_2v_2$ since W is a subspace. It follows from induction that any linear combination $c_1v_1 + c_2v_2 + ... + c_nv_n$ is in W.

Thus $Span(v_1, v_2, \dots, v_n) \subset W$

# Ex 6: Given subspaces U, V of $R^n$, $U \cap V$ and $U+V$ are subspaces, but $U \cup V$ is not.

(a) Let $W = U \cap V = \{\vec{x} \in R^n: x \in U and x \in V\}$ is also a subspace.

Proof: 
- 0 is in U and V
- if x is in W, then cx is in U and V, hence in W
- if x and y are in w, x + y are in U and V, hence in W

An example is 2 planes intersecting. For example Span(e_1, e_2) and Span(e_2,e_3) gives Span(e_2) as a subspace.

(b) Let $W = U \cup V = \{\vec{x} \in R^n: x \in U or x \in V\}$

See for example the intersection of Span($e_1$) and Span($e_2$). $e_1 + e_2$ is not in the union.

(c) Let $W = U + V = \{\vec{x} \in R^n: x = u + w \text{ for some u } \in U \text{ and some } v \in V\}$ is also a subspace.

Proof: 
- 0 is in W
- If x is in W, then x = u + v where u is in U and v is in V. Then cx = cu + cv is in W since cu is in U, and cv is in V.
- if x and y are in W, x = u + v, y = u' + v' hence x + y = (u + u') + (v + v') is in W since u+u' is in U, and v+v' is in V.


# Ex 7: $Span(v_1, v_2, \dots, v_k) = Span(v_1, v_2, \dots, v_k, v) \iff v \in Span(v_1, v_2, \dots, v_k)$

a) Given Span(v_1, v_2, \dots, v_k) = Span(v_1, v_2, \dots, v_k, v)

Obviously v is in $Span(v_1, v_2, \dots, v_k, v)$ and hence in $Span(v_1, v_2, \dots, v_k)$

b) Given $v \in Span(v_1, v_2, \dots, v_k)$:

Clearly $Span(v_1, v_2, \dots, v_k) \subset Span(v_1, v_2, \dots, v_k, v)$. We just have to show:

$Span(v_1, v_2, \dots, v_k, v) \subset Span(v_1, v_2, \dots, v_k)$

Let $v = d_1v_1 + d_2v_2 + \dots + d_kv_k$ (by hypothesis that v is in the span).

Let $x \in Span(v_1, v_2, \dots, v_k, v)$ then $x = c_1v_1 + c_2v_2 + \dots + c_kv_k + cv = (c_1 + d_1)v_1 + (c_2 + d_2)v_2 + \dots + (c_k + d_k)v_k$.

Thus $x \in Span(v_1, v_2, \dots, v_k)$







# Ex 8: Let V be a subspace in $R^n$. Prove $V \cap V^{\perp} = \{\vec{0}\}$

Proof:

- 0 is in V and $V^{\perp}$, so in $V \cap V^{\perp}$
- Let x (non-zero) be in V. Then x.x is non-zero, so x is not in $V^{\perp}$
- If y (non-zero) is in both V and $V^{\perp}$, then y.y = 0 => y = 0, a contradiction.




# Ex 9: Let U and V be subspaces, such that $U \subset V$. Then $V^{\perp} \subset U^{\perp}$

Proof:

For any x in $V^{\perp}$, x.u = 0 for any u in U, since $U \subset V$. It follows that $V^{\perp} \subset U^{\perp}$.

# Ex 10: Let $V \in R^n$ be a subspace. Then $V \subset (V^{\perp})^{\perp}$

Let x be any vector in $V^{\perp}$ and v in V. Then x.v = 0 => $v \in (V^{\perp})^{\perp}$.

In fact, logically, any non-zero vector not in V or $V^{\perp}$ should be expressable as a sum of components of both subspaces. However, this is not proven here.

# Ex 11: Let V = $Span(v_1, v_2, \dots, v_k)$ in $R^n$. Show that there are mutually orthogonal vectors $w_1, w_2, \dots, w_k$ in V, such that Span($w_1, w_2, \dots, w_k$) = V

Proof: Basically, let $w_1 = v_1$, $w_2 = v_2 - \frac{v_2.w_1}{\Vert w_1 \Vert^2} w_1$, $w_k = v_k - \sum_{i=1}^{k-1} \frac{v_k.w_i}{\Vert w_i \Vert^2} w_i$

Let P(n) be the statement that vectors $w_1, w_2, ..., w_n$ are mutually orthogonal, and span($w_1, w_2, ..., w_n$) = span($v_1, v_2, ..., v_n$)

a) Prove P(2) is true

Clearly $w_1.w_2 = v_1.w_2 = 0$ since $w_2$ is just the perpendicular component of $v_2$ w.r.t $v_1$.

We can see that $w_1, w_2$ are linear combinations of $v_1, v_2$, so $span(w_1, w_2) \subset span(v_1, v_2)$

Similarly, we can express $v_1, v_2$ as linear combination of $w_1, w_2$, hence $span(v_1, v_2) \subset span(w_1, w_2)$

b) Assume P(k) is true. i.e. $w_1, w_2, ..., w_k$ are mutually orthogonal and span($w_1, w_2, ..., w_k$) = span($v_1, v_2, ..., v_k$)

c) Now given P(k), we have to show P(k+1) is true: Let

$w_{k+1} = v_{k+1} - \sum_{i=1}^{k} \frac{v_{k+1}.w_i}{\Vert w_i \Vert^2} w_i$

Note that given P(k) is true in terms of span equality, we can express:

$\sum_{i=1}^{k} \frac{v_{k+1}.w_i}{\Vert w_i \Vert^2} w_i = \sum_{i=1}^{k} e_i v_i$


Taking the dot product with $w_i$, i between 1 and k, we see:
$w_{k+1} \cdot w_i = v_{k+1}.w_i - \frac{v_{k+1}.w_i}{\Vert w_i \Vert^2} w_i.w_i = 0$ 



This is because given P(k) is true, $w_i.w_j = 0$ whenever $i \ne j$, $i,j \le k$
So $w_{k+1}$ is orthogonal to $w_1, w_2, ..., w_k$.


Now given $x \in span(w_1, w_2, ..., w_k, w_{k+1})$, 

$$
\begin{align}

x &= c_1w_1 + c_2w_2 + \dots + c_kw_k + c_{k+1}w_{k+1} \\
  &= d_1v_1 + d_2v_2 + \dots + d_kv_k + c_{k+1}(v_{k+1} - \sum_{i=1}^{k} \frac{v_{k+1}.w_i}{\Vert w_i \Vert^2} w_i) \text{ because of P(k) } \\
  &= d_1v_1 + d_2v_2 + \dots + d_kv_k + c_{k+1}(v_{k+1} - \sum_{i=1}^{k} e_i v_i) \\
  &= \sum_{i=1}^{k} (d_i - e_i) v_i + c_{k+1}v_{k+1}

\end{align}
$$

Thus $x \in span(v_1, v_2, ..., v_k, v_{k+1})$

Thus P(k+1) is true, completing the proof.
